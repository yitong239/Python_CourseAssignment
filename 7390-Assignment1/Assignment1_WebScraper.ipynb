```python
html ="""
<!DOCTYPE html><html lang="en"><head><meta charset="utf-8"/><meta http-equiv="X-UA-Compatible" content="IE=edge"/><title>The First Time I Met Paul Bocuse | The New Yorker</title><meta name="msapplication-tap-highlight" content="no"/><meta name="viewport" content="user-scalable=no,width=device-width,initial-scale=1,maximum-scale=1"/><link rel="canonical" href="https://www.newyorker.com/culture/annals-of-gastronomy/the-first-time-i-met-paul-bocuse"/><link rel="shortcut icon" type="image/x-icon" href="/favicon.ico"/><meta name="id" content="5a6602e016b0267062e903f3"/><script>(function(d)
  { var e = d.createElement('script'); e.src = d.location.protocol + '//tag.bounceexchange.com/1990/i.js'; e.async = true; d.getElementsByTagName("head")[0].appendChild(e); }
  (document));</script><script type="application/ld+json">{"@context":"http://schema.org","@type":"NewsArticle","headline":"The First Time I Met Paul Bocuse","alternativeHeadline":"The First Time I Met Paul Bocuse","articleBody":"","articleSection":"Annals of Gastronomy","dateModified":"2018-01-23T00:37:42.779Z","datePublished":"2018-01-23T00:08:28.877Z","description":"","isAccessibleForFree":false,"keywords":["food","postscript","cooking","france"],"thumbnailUrl":"https://media.newyorker.com/photos/5a6633cd1ccea91635374116/4:3/w_960,c_limit/buford-paul-bocuse-2.jpg","url":"https://www.newyorker.com/culture/annals-of-gastronomy/the-first-time-i-met-paul-bocuse","author":{"@context":"http://schema.org","@type":"Person","name":["Bill Buford"]},"creator":{"@context":"http://schema.org","@type":"Person","name":["Bill Buford"]},"image":{"@context":"http://schema.org","@type":"ImageObject","url":"https://media.newyorker.com/photos/5a6633cd1ccea91635374116/4:3/w_960,c_limit/buford-paul-bocuse-2.jpg","width":0,"height":0},"publisher":{"@context":"http://schema.org","@type":"Organization","name":"The New Yorker","url":"https://www.newyorker.com","logo":{"@context":"http://schema.org","@type":"ImageObject","url":"https://www.newyorker.com/images/svg/tny-logo.svg","width":190,"height":44}},"mainEntityOfPage":{"@context":"http://schema.org","@type":"WebPage","@id":"https://www.newyorker.com/culture/annals-of-gastronomy/the-first-time-i-met-paul-bocuse"}}</script><script type="text/javascript" id="evolok-include">var EVOLOK = {
                "activeMeter"    : {
                    "metername"  : "dft",
                    "max"        : "4",
                    "count"      : "1",
                    "viewed"     : "0",
                    "abGroup"    : "4MeterMax",
                    "abValue"    : "1"
                },
                "cnee"             : {
                    "subscription" : "",
                    "entitlement"  : ""
                },
                 "sessionInfo"    : {
                    "ev_sid"     : "5a66a616e4b0da4070bbbf2d",
                    "ev_did"     : "5a66a616e4b0da4070bbbf2a",
                    "ev_ss"      : ""
                },
                "exceededMeter"  : "0",
                "pid"            : "",
                "paywall_status" : "ALLOW_ACCESS"
            };</script><meta name="parsely-metadata" content="{&quot;id&quot;:&quot;5a6602e016b0267062e903f3&quot;,&quot;issueDate&quot;:&quot;&quot;,&quot;dek&quot;:&quot;Bocuse, who died on Saturday, at the age of ninety-one, had an undeniably infectious culinary charisma. He was what people want a grand French\nchef to be. &quot;,&quot;section&quot;:&quot;Annals of Gastronomy&quot;,&quot;url&quot;:&quot;https://www.newyorker.com/culture/annals-of-gastronomy/the-first-time-i-met-paul-bocuse&quot;,&quot;image&quot;:{&quot;@type&quot;:&quot;ImageObject&quot;,&quot;contentUrl&quot;:&quot;https://www.newyorker.com/culture/annals-of-gastronomy/the-first-time-i-met-paul-bocuse&quot;,&quot;url&quot;:&quot;https://media.newyorker.com/photos/5a6633cd1ccea91635374116/4:3/w_960,c_limit/buford-paul-bocuse-2.jpg&quot;}}"/><link rel="amphtml" href="https://www.newyorker.com/culture/annals-of-gastronomy/the-first-time-i-met-paul-bocuse/amp"/><meta name="copyright" content="Copyright (c) Condé Nast 2018"/><meta name="description" content="Bill Buford remembers the French chef Paul Bocuse, who died on Saturday, at the age of ninety-one."/><meta name="keywords" content="food,postscript,cooking,france"/><meta name="apple-itunes-app" content="app-id=1081530898"/><meta property="fb:pages"/><meta property="fb:app_id" content="1147169538698836"/><meta name="news_keywords" content="Culture, Annals of Gastronomy, food, postscript, cooking, france"/><meta property="og:locale" content="en_US"/><meta property="og:site_name" content="The New Yorker"/><meta property="og:image" content="https://media.newyorker.com/photos/5a6633cd1ccea91635374116/16:9/w_1200,h_630,c_limit/buford-paul-bocuse-2.jpg"/><meta property="og:title" content="The First Time I Met Paul Bocuse"/><meta property="og:description" content="Bocuse, who died on Saturday, at the age of ninety-one, had an undeniably infectious culinary charisma. He was what people want a grand French
chef to be."/><meta property="og:url" content="https://www.newyorker.com/culture/annals-of-gastronomy/the-first-time-i-met-paul-bocuse"/><meta property="og:type" content="article"/><meta name="twitter:site" content="@NewYorker"/><meta name="twitter:domain" content="www.newyorker.com"/><meta name="twitter:card" content="summary_large_image"/><meta name="twitter:title" content="The First Time I Met Paul Bocuse"/><meta name="twitter:description" content="Bocuse, who died on Saturday, at the age of ninety-one, had an undeniably infectious culinary charisma. He was what people want a grand French
chef to be."/><meta name="twitter:image:src" content="https://media.newyorker.com/photos/5a6633cd1ccea91635374116/16:9/w_1200,h_630,c_limit/buford-paul-bocuse-2.jpg?mbid=social_retweet"/><meta name="pubdate" content="20180123"/><script>window.__TNY__=window.__TNY__||{};</script><script>window.__AUTOPILOT__={"analytics:parsely:key":"newyorker.com","env":"production","vulcan:host":"media.newyorker.com","copilot:host":"tny-api.aws.conde.io","analytics:simplereach:id":"543eb257d70ba089a3000004","analytics:google:id":"UA-8293713-14","analytics:omniture:name":"newyorker","autopilot:host":"www.newyorker.com","brand:humanName":"The New Yorker","brand:siteCode":"tny"}||{};</script><script>!function (e){"use strict";var t=e||{};window.cns=window.cns||{async:function(e,t){this.queue[e]||(this.queue[e]=[]),"function"==typeof t&&this.queue[e].push(t)},queue:{},embed:{backup:t.overwrite}};var n=document.createElement("script");t.isAsync?n.async=!0:n.defer=!0,n.src=function(){if(t.overwrite)return t.overwrite;var e=location.hostname,n=""!==location.port?":"+location.port:"",o=location.search.match(/CNS_SB=([^&\s]+)[&\s]?/)||[],s=o.length?"-sandbox/"+o[1]:"",c=!t.bucketer||s?"":t.bucketer(),i="";if(!1!==t.bustCache){var r=new Date;r.setMinutes(Math.floor(r.getMinutes()/15),0,0),i+="?"+r.getTime()}return"//"+(e+n)+"/cns"+s+c+"/services.min.js"+i}(),document.head.appendChild(n)}({});</script><link rel="stylesheet" href="/css/article.css?cb=8d41d241"/></head><body class="page-article"><div id="app-root"><div data-reactroot="" data-reactid="1" data-react-checksum="-637133841"><div class="" data-reactid="2"><main class="PageContainer__main___w2VEm" data-reactid="3"><div class="SiteHeader__siteHeader___2Fkqk" data-reactid="4"><div class="SiteHeader__navWrapper___1FuV4 " data-reactid="5"><header class="MainHeader__mainHeader___28Tsw MainHeader__partial___cL3c0" data-reactid="6"><div class="MainHeader__topBar___1yqIc" data-reactid="7"><div class="MainHeader__topBarItems___13CTQ" data-reactid="8"><div class="MainHeader__subscribeLink___3TGEc" data-reactid="9"><div class="MainHeader__fullSubscribe___dolgp" data-reactid="10"><!-- react-empty: 11 --></div><div class="MainHeader__collapseSubscribe___2doFO" data-reactid="12"><!-- react-empty: 13 --></div></div><div data-track-type="nav" data-track-location="logo" class="Logo__componentLogo___1mTF4 Logo__partial___2XOqH" data-reactid="14"><a class="Logo__logo___3DzHd" href="/" data-reactid="15"><svg viewBox="0 0 282 66" xmlns="http://www.w3.org/2000/svg" role="img" aria-labelledby="tnyLogo"><title id="tnyLogo">The New Yorker</title><g fill="none" fill-rule="evenodd"><path fill="#121212" d="M.4 24.6h4.8L25.4 50V24.5h3V64l-25-31.8v31.4h-3zm33.4 0h14.5l2 4.5s-2.4-2.3-7-2.3H40v13.8h5.7c3 0 3.5-.4 4.6-1.5v5.8s-1.3-2-4.6-2H40v18h8c8 0 12-3 12-3l-3 5.8H33.8v-39zm200.5 0h16l2 4.5s-2.2-2.3-7-2.3h-5.4v13.8h6.2c2.4 0 3-.2 4-1V44s-.7-1.4-4-1.4H240V61h3.8c7 0 9.6-2 9.6-2l-3.5 4.6h-15.7v-39zM52 24.6h6.2L71.4 53l2.7-5.8-10-22.6h5.8L77 39.8l7-15.2h3L79 44l4.3 9L96 24.7h3.5L81.8 64l-6-13-6 13zm102.3-1.4c13.7 0 22 9.5 22 20 0 13.6-7.5 22.2-21.6 22.2-14.2 0-22.6-10.5-22.6-21 0-11.8 9-21.2 22.3-21.2zM138 43.5c0 9.2 5.2 19.2 17 19.2 11 0 15.3-9 15.3-19 0-8-5.5-18.2-16.3-18.2-7.4 0-16 5.8-16 18zm43-19h9c5 0 9 2.8 9 8 0 6-4.4 7.7-7.5 8.7 5.6 1 7.7 3.6 7.7 11 0 7.5 2 10.2 3.6 10.2 1.3 0 2.4-.8 2.7-1.4l.8.5c-.3.4-1.8 2.8-5.4 2.8s-6.5-3.3-6.5-11c0-11.3-3-11.5-5.5-11.5V41s5.2-1.2 5.2-7.5c0-6-3.6-6.8-5-6.8h-2.7v37h-5.8l.2-39zm75 0h9.2c5 0 9 2.8 9 8 0 6-4.5 7.7-7.6 8.7 5.6 1 7.7 3.6 7.7 11 0 7.5 2 10.2 3.6 10.2 1.2 0 2.3-.8 2.6-1.4l.8.5c-.3.4-1.8 2.8-5.4 2.8s-6.4-3.3-6.4-11c0-11.3-3-11.5-5.5-11.5V41c.2 0 5.3-1.2 5.3-7.5 0-6-3.6-6.8-5-6.8h-2.7v37h-5.8l.2-39zm-31 27.6c0 7.6 1.8 10.3 3.6 10.3 1.3 0 2.3-1 2.7-1.4l.8.4c-.2.3-1.7 2.7-5.4 2.7-3.6 0-6.3-3-6.3-11v-1.5c0-4.6-.3-8.2-1-10.5-.7-2-3-3.3-4-3.3h-1v25.8h-5.8v-39h5.7v11.4l10.3-11.6h3.7l-11 12.2c5.7 1 7.7 5 7.7 14V52zM135.5.4h3v7h7v-7h3v16.3h-3v-8h-7v8h-3zm18.6 0h7.6l1 2.4s-1.6-1-3-1h-2.7v5.6h2.6c1.4 0 1.8-.6 1.8-.6v2.6s-.4-.8-1.8-.8H157v7h3.6c2.7 0 3.8-1.4 3.8-1.4l-1.3 2.7h-9V.3h.2zM126 1.6h1.4c2.5 0 4.5 1.7 4.5 1.7l-1-3h-13.5l-1 3s2-1.7 4.6-1.7h1.7v15h3v-15h.2z"/><path fill="#000" d="M124 63.6V43.4l11-18.8h-3.6l-10.4 18-10.5-18H107l11 18.8v20.2z"/></g></svg></a></div><div class="UserInfo__userInfo___nvFKn UserInfo__partial___3-Dk_ undefined" data-reactid="16"><div class="UserInfo__userInfoItems___2TxZh" data-reactid="17"><a class="Link__link___3dWao UserInfo__shop___20W-I " href="http://www.condenaststore.com/~/new-yorker-store?utm_medium=referral&amp;utm_source=NewYorker&amp;utm_content=SHOP&amp;AID=1247905545" data-reactid="18"><!-- react-text: 19 -->Shop<!-- /react-text --></a><div class="SignIn__signIn___1aO_Z" data-reactid="20"><a class="Link__link___3dWao  " href="https://account.newyorker.com" data-reactid="21"><!-- react-text: 22 -->Sign in<!-- /react-text --></a><span data-reactid="23">|</span><a class="Link__link___3dWao  " href="https://account.newyorker.com/linksubscription" data-reactid="24"><!-- react-text: 25 -->Link your subscription<!-- /react-text --></a></div></div><div data-track-type="nav" data-track-location="search-icon" class="UserInfo__search___1oSUo" data-reactid="26"><div class="Lightbox__lightbox___2lLZl Lightbox__white___jj_9p  " role="button" tabindex="0" data-reactid="27"><div class="" data-reactid="28"><a class="Link__link___3dWao SearchModal__icon___2cTZE " data-reactid="29"><div class="Link__content___6CYL1" data-reactid="30"><i class="Link__icon___3s7us Link__left___1QLUQ " data-reactid="31"><svg viewBox="0 0 18 18" xmlns="http://www.w3.org/2000/svg"><path d="M2 7.5C2 4.468 4.468 2 7.5 2S13 4.468 13 7.5 10.532 13 7.5 13 2 10.532 2 7.5zm16 9.086l-4.545-4.545C14.42 10.779 15 9.208 15 7.5 15 3.364 11.636 0 7.5 0S0 3.364 0 7.5 3.364 15 7.5 15c1.708 0 3.279-.58 4.541-1.545L16.586 18 18 16.586z" fill="#121212" fill-rule="evenodd" /></svg></i></div></a></div></div></div></div></div></div><nav data-reactid="32"><div class="Navigation__navList___gJe7C Navigation__partial___1uoe1 undefined" data-reactid="33"><ul data-reactid="34"><li data-track-type="nav" data-track-location="button-news" data-reactid="35"><a class="Link__link___3dWao  " href="/news" data-reactid="36"><!-- react-text: 37 -->News<!-- /react-text --></a></li><li data-track-type="nav" data-track-location="button-culture" data-reactid="38"><a class="Link__link___3dWao  " href="/culture" data-reactid="39"><!-- react-text: 40 -->Culture<!-- /react-text --></a></li><li data-track-type="nav" data-track-location="button-books" data-reactid="41"><a class="Link__link___3dWao  " href="/books" data-reactid="42"><!-- react-text: 43 -->Books<!-- /react-text --></a></li><li data-track-type="nav" data-track-location="button-business-&amp;-tech" data-reactid="44"><a class="Link__link___3dWao  " href="/business" data-reactid="45"><!-- react-text: 46 -->Business &amp; Tech<!-- /react-text --></a></li><li data-track-type="nav" data-track-location="button-humor" data-reactid="47"><a class="Link__link___3dWao  " href="/humor" data-reactid="48"><!-- react-text: 49 -->Humor<!-- /react-text --></a></li><li data-track-type="nav" data-track-location="button-cartoons" data-reactid="50"><a class="Link__link___3dWao  " href="/cartoons" data-reactid="51"><!-- react-text: 52 -->Cartoons<!-- /react-text --></a></li><li data-track-type="nav" data-track-location="button-magazine" data-reactid="53"><a class="Link__link___3dWao  " href="/magazine" data-reactid="54"><!-- react-text: 55 -->Magazine<!-- /react-text --></a></li><li data-track-type="nav" data-track-location="button-video" data-reactid="56"><a class="Link__link___3dWao  " href="http://video.newyorker.com" data-reactid="57"><!-- react-text: 58 -->Video<!-- /react-text --></a></li><li data-track-type="nav" data-track-location="button-podcasts" data-reactid="59"><a class="Link__link___3dWao  " href="/podcast" data-reactid="60"><!-- react-text: 61 -->Podcasts<!-- /react-text --></a></li><li data-track-type="nav" data-track-location="button-archive" data-reactid="62"><a class="Link__link___3dWao  " href="/archive" data-reactid="63"><!-- react-text: 64 -->Archive<!-- /react-text --></a></li><li data-track-type="nav" data-track-location="button-goings-on" data-reactid="65"><a class="Link__link___3dWao  " href="/goings-on-about-town/" data-reactid="66"><!-- react-text: 67 -->Goings On<!-- /react-text --></a></li><li class="Navigation__subscribe___3OCs_" data-reactid="68"><!-- react-empty: 69 --><div class="DropDown__dropDown___2J7gQ Navigation__showDropDown___2BXqe" data-reactid="70"><div class="DropDown__hoverSpace___2KGci" data-reactid="71"></div><!-- react-empty: 72 --></div></li></ul></div></nav></header><header class="MobileHeader__mobileHeader___13HZG" data-reactid="73"><button class="MobileHeader__hamburger___3x1fn" data-reactid="74"><div class="MobileHeader__middle___3eg8Y" data-reactid="75"></div></button><div data-track-type="nav" data-track-location="logo" class="Logo__componentLogo___1mTF4 Logo__mobileHeader___7s-0k" data-reactid="76"><a class="Logo__logo___3DzHd" href="/" data-reactid="77"><svg viewBox="0 0 282 66" xmlns="http://www.w3.org/2000/svg" role="img" aria-labelledby="tnyLogo"><title id="tnyLogo">The New Yorker</title><g fill="none" fill-rule="evenodd"><path fill="#121212" d="M.4 24.6h4.8L25.4 50V24.5h3V64l-25-31.8v31.4h-3zm33.4 0h14.5l2 4.5s-2.4-2.3-7-2.3H40v13.8h5.7c3 0 3.5-.4 4.6-1.5v5.8s-1.3-2-4.6-2H40v18h8c8 0 12-3 12-3l-3 5.8H33.8v-39zm200.5 0h16l2 4.5s-2.2-2.3-7-2.3h-5.4v13.8h6.2c2.4 0 3-.2 4-1V44s-.7-1.4-4-1.4H240V61h3.8c7 0 9.6-2 9.6-2l-3.5 4.6h-15.7v-39zM52 24.6h6.2L71.4 53l2.7-5.8-10-22.6h5.8L77 39.8l7-15.2h3L79 44l4.3 9L96 24.7h3.5L81.8 64l-6-13-6 13zm102.3-1.4c13.7 0 22 9.5 22 20 0 13.6-7.5 22.2-21.6 22.2-14.2 0-22.6-10.5-22.6-21 0-11.8 9-21.2 22.3-21.2zM138 43.5c0 9.2 5.2 19.2 17 19.2 11 0 15.3-9 15.3-19 0-8-5.5-18.2-16.3-18.2-7.4 0-16 5.8-16 18zm43-19h9c5 0 9 2.8 9 8 0 6-4.4 7.7-7.5 8.7 5.6 1 7.7 3.6 7.7 11 0 7.5 2 10.2 3.6 10.2 1.3 0 2.4-.8 2.7-1.4l.8.5c-.3.4-1.8 2.8-5.4 2.8s-6.5-3.3-6.5-11c0-11.3-3-11.5-5.5-11.5V41s5.2-1.2 5.2-7.5c0-6-3.6-6.8-5-6.8h-2.7v37h-5.8l.2-39zm75 0h9.2c5 0 9 2.8 9 8 0 6-4.5 7.7-7.6 8.7 5.6 1 7.7 3.6 7.7 11 0 7.5 2 10.2 3.6 10.2 1.2 0 2.3-.8 2.6-1.4l.8.5c-.3.4-1.8 2.8-5.4 2.8s-6.4-3.3-6.4-11c0-11.3-3-11.5-5.5-11.5V41c.2 0 5.3-1.2 5.3-7.5 0-6-3.6-6.8-5-6.8h-2.7v37h-5.8l.2-39zm-31 27.6c0 7.6 1.8 10.3 3.6 10.3 1.3 0 2.3-1 2.7-1.4l.8.4c-.2.3-1.7 2.7-5.4 2.7-3.6 0-6.3-3-6.3-11v-1.5c0-4.6-.3-8.2-1-10.5-.7-2-3-3.3-4-3.3h-1v25.8h-5.8v-39h5.7v11.4l10.3-11.6h3.7l-11 12.2c5.7 1 7.7 5 7.7 14V52zM135.5.4h3v7h7v-7h3v16.3h-3v-8h-7v8h-3zm18.6 0h7.6l1 2.4s-1.6-1-3-1h-2.7v5.6h2.6c1.4 0 1.8-.6 1.8-.6v2.6s-.4-.8-1.8-.8H157v7h3.6c2.7 0 3.8-1.4 3.8-1.4l-1.3 2.7h-9V.3h.2zM126 1.6h1.4c2.5 0 4.5 1.7 4.5 1.7l-1-3h-13.5l-1 3s2-1.7 4.6-1.7h1.7v15h3v-15h.2z"/><path fill="#000" d="M124 63.6V43.4l11-18.8h-3.6l-10.4 18-10.5-18H107l11 18.8v20.2z"/></g></svg></a></div><div class="MobileHeader__mobileSubscribe___3OHUs" data-reactid="78"><!-- react-empty: 79 --></div></header></div><div class="SiteHeader__secondaryNav___20mer SiteHeader__hideSecondary___w4vbX" data-reactid="80"><span class="SiteHeader__eustace___2yJ2Y" data-reactid="81"><svg version="1.1" viewBox="0 0 40 40" xmlns="http://www.w3.org/2000/svg" xmlns:xlink="http://www.w3.org/1999/xlink"><g transform="translate(-14 -12)" fill="none" fill-rule="evenodd"><image x="14" y="12" width="40" height="40" xlink:href=""/></g></svg></span><p class="SiteHeader__title___aZr3Q" data-reactid="82"></p></div></div><section class="PageContainer__pageContent___1xERg PageContainer__article___136yK" data-reactid="83"><!-- react-empty: 84 --><article class="Layout__layoutContainer___2gtig" data-reactid="85"><header data-reactid="86"><header class="ArticleHeader__articleHeader___1G7-9  ArticleHeader__default___1GpE3" data-reactid="87"><hgroup class="ArticleHeader__headerRow___nDCwd" data-reactid="88"><div class="ArticleHeader__rubricAndIssue___1YUtt" data-reactid="89"><div class="ArticleHeader__rubric___3YLRT" data-reactid="90"><a class="Link__link___3dWao Rubric__rubric___3hrQE  " href="/culture/annals-of-gastronomy" title="Annals of Gastronomy" data-reactid="91"><!-- react-text: 92 -->Annals of Gastronomy<!-- /react-text --></a></div></div><h1 class="ArticleHeader__hed___GPB7e" data-reactid="93">The First Time I Met Paul Bocuse</h1><!-- react-text: 94 --><!-- /react-text --><div class="ArticleHeader__metaInfo___1aBON" data-reactid="95"><div class="ArticleContributors__byline___3-luq" data-reactid="96"><div class="ArticleContributors__contributorWrapper___1CrIJ" data-reactid="97"><div class="Byline__articleHeader___13Q7D " data-reactid="98"><p class="Byline__by___37lv8" data-reactid="99"><!-- react-text: 100 -->By <!-- /react-text --><a class="Link__link___3dWao  " title="Bill Buford" rel="author" href="/contributors/bill-buford" data-reactid="101"><!-- react-text: 102 -->Bill Buford<!-- /react-text --></a></p><p class="ArticleTimestamp__timestamp___1klks " data-reactid="103">7:08 P.M.</p></div></div></div><ul class="ArticleSocial__socialShare___2Skji ArticleSocial__header___1huCC" data-reactid="104"><li data-track-type="socialShare" data-track-location="default" data-source="facebook" class="SocialShareItem__icon___3QQZ0 SocialShareItem__facebook___15Vks " data-reactid="105"><a class="Link__link___3dWao  " href="https://www.facebook.com/sharer/sharer.php?u=https%3A%2F%2Fwww.newyorker.com%2Fculture%2Fannals-of-gastronomy%2Fthe-first-time-i-met-paul-bocuse&amp;display=popup&amp;ref=plugin" data-reactid="106"><div class="Link__content___6CYL1" data-reactid="107"><i class="Link__icon___3s7us Link__left___1QLUQ " data-reactid="108"><svg width="8" height="17" viewBox="0 0 8 17" xmlns="http://www.w3.org/2000/svg"><path d="M1.7323 17H5.2V8.499h2.3135L7.82 5.5698H5.2l.0037-1.4667c0-.7634.072-1.1733 1.154-1.1733h1.4458V0h-2.314C2.7104 0 1.7324 1.4214 1.7324 3.811v1.7588H0v2.9297h1.7323V17z" fill="#000" fill-rule="evenodd"/></svg></i></div></a></li><li data-track-type="socialShare" data-track-location="default" data-source="twitter" class="SocialShareItem__icon___3QQZ0 SocialShareItem__twitter___1KPYQ " data-reactid="109"><a class="Link__link___3dWao  " href="https://twitter.com/intent/tweet?original_referer=https%3A%2F%2Fwww.newyorker.com%2Fculture%2Fannals-of-gastronomy%2Fthe-first-time-i-met-paul-bocuse&amp;text=The%20First%20Time%20I%20Met%20Paul%20Bocuse&amp;tw_p=tweetbutton&amp;url=https%3A%2F%2Fwww.newyorker.com%2Fculture%2Fannals-of-gastronomy%2Fthe-first-time-i-met-paul-bocuse" data-reactid="110"><div class="Link__content___6CYL1" data-reactid="111"><i class="Link__icon___3s7us Link__left___1QLUQ " data-reactid="112"><svg width="19" height="15" viewBox="0 0 19 15" xmlns="http://www.w3.org/2000/svg"><path d="M18.223 1.739c-.654.295-1.3565.4943-2.094.584.7527-.459 1.3308-1.1858 1.603-2.0518-.7045.425-1.4847.7336-2.3152.9C14.7518.4504 13.8043 0 12.7558 0c-2.0137 0-3.646 1.6605-3.646 3.7084 0 .2908.032.5737.0943.8452-3.03-.1547-5.7164-1.631-7.5146-3.8748-.3138.5478-.4936 1.1848-.4936 1.8644 0 1.2866.6436 2.4217 1.622 3.0868-.5977-.0192-1.16-.186-1.6515-.464l-.0003.0468c0 1.7968 1.2567 3.2957 2.9246 3.6364-.306.0847-.628.13-.9605.13-.235 0-.4632-.0232-.686-.0665.464 1.4733 1.8105 2.5455 3.406 2.5755-1.248.9946-2.82 1.5875-4.528 1.5875-.2943 0-.5845-.0176-.8698-.0518 1.6136 1.052 3.53 1.666 5.5888 1.666 6.706 0 10.3733-5.651 10.3733-10.5515 0-.1608-.0036-.3208-.0106-.4798.7123-.5228 1.3304-1.176 1.8192-1.9197" fill="#000" fill-rule="evenodd"/></svg></i></div></a></li><li data-track-type="socialShare" data-track-location="default" data-source="email" class="SocialShareItem__icon___3QQZ0 SocialShareItem__email___aXa4V " data-reactid="113"><a class="Link__link___3dWao  " href="mailto:?subject=From%20newyorker.com:%20The%20First%20Time%20I%20Met%20Paul%20Bocuse&amp;body=Bocuse%2C%20who%20died%20on%20Saturday%2C%20at%20the%20age%20of%20ninety-one%2C%20had%20an%20undeniably%20infectious%20culinary%20charisma.%20He%20was%20what%20people%20want%20a%20grand%20French%0Achef%20to%20be.%20%0D%0Ahttps%3A%2F%2Fwww.newyorker.com%2Fculture%2Fannals-of-gastronomy%2Fthe-first-time-i-met-paul-bocuse" data-reactid="114"><div class="Link__content___6CYL1" data-reactid="115"><i class="Link__icon___3s7us Link__left___1QLUQ " data-reactid="116"><svg width="19" height="14" viewBox="0 0 19 14" xmlns="http://www.w3.org/2000/svg"><path d="M9.7902 9.4853l-.003-.003-.004.003-6.9808-7.071 1.396-1.414 5.5888 5.66 5.588-5.66 1.396 1.414-6.981 7.071zm-8.9624 4.515h17.7707v-14H.8278v14z" fill="#000" fill-rule="evenodd"/></svg></i></div></a></li><li data-track-type="socialShare" data-track-location="default" data-source="print" class="SocialShareItem__icon___3QQZ0 SocialShareItem__print___2zPf0 " data-reactid="117"><a class="Link__link___3dWao  " href="#" data-reactid="118"><div class="Link__content___6CYL1" data-reactid="119"><i class="Link__icon___3s7us Link__left___1QLUQ " data-reactid="120"><svg width="17" height="17" viewBox="0 0 17 17" xmlns="http://www.w3.org/2000/svg"><g fill="#FFF" fill-rule="evenodd"><path d="M4.9688 1.988h8.008v2.009h.999V1.988c0-.559-.456-.999-.999-.999h-8.008c-.559 0-.999.456-.999.999v2.009h.999V1.988zM14.9857 8.0037c-.559 0-.999-.456-.999-.999 0-.559.456-.999.999-.999.559 0 .999.456.999.999.016.543-.44.999-.999.999zm-1.993 8.008h-8.024v-7.009h8.008v7.015h.016v-.006zm3.008-11.016H1.9777c-.56 0-.999.456-.999.999v5.001c0 .559.456.998.999.998h2.008v4.007c0 .559.456.999.999.999h8.008c.56 0 .999-.456.999-.999v-3.985h2.009c.559 0 .999-.456.999-.999v-5.022c0-.543-.456-.999-.999-.999z"/><path d="M5.979 11.012h4.007v-.999H5.979M5.979 13.01h6.016v-.999H5.979M5.979 15.018h6.016v-.999H5.979"/></g></svg></i></div></a></li></ul></div></hgroup></header></header><div class="Layout__twoColumn___1sIWV" data-reactid="121"><main class="Layout__content___5vVe9" data-reactid="122"><div data-reactid="123"><div class="ArticleLedeImage__fullWidth___JZlQO " data-reactid="124"><div class="ArticleLedeImage__container___Fy9Ni" data-reactid="125"><div class="Lightbox__lightbox___2lLZl Lightbox__white___jj_9p  " role="button" tabindex="0" data-reactid="126"><figure class="Figure__figure___U_9Te Figure__fullHeight___3uICS ArticleLedeImage__lede___1rVAF " data-reactid="127"><div class="placeholder" data-reactid="128"><div class="placeholder-buttress" data-reactid="129"></div><div class="placeholder-content" data-reactid="130"><div class="Image__image___1PhYl Figure__image___1hDvt ArticleLedeImage__image___17_0r" role="false" tabindex="false" data-reactid="131"><picture class="component-responsive-image" data-reactid="132"><source media="(min-width: 1280px)" srcset="https://media.newyorker.com/photos/5a6633cd1ccea91635374116/master/w_649,c_limit/buford-paul-bocuse-2.jpg, https://media.newyorker.com/photos/5a6633cd1ccea91635374116/master/w_1298,c_limit/buford-paul-bocuse-2.jpg 2x" data-reactid="133"/><source media="(min-width: 1024px)" srcset="https://media.newyorker.com/photos/5a6633cd1ccea91635374116/master/w_813,c_limit/buford-paul-bocuse-2.jpg, https://media.newyorker.com/photos/5a6633cd1ccea91635374116/master/w_1626,c_limit/buford-paul-bocuse-2.jpg 2x" data-reactid="134"/><source media="(min-width: 768px)" srcset="https://media.newyorker.com/photos/5a6633cd1ccea91635374116/master/w_727,c_limit/buford-paul-bocuse-2.jpg, https://media.newyorker.com/photos/5a6633cd1ccea91635374116/master/w_1454,c_limit/buford-paul-bocuse-2.jpg 2x" data-reactid="135"/><source srcset="https://media.newyorker.com/photos/5a6633cd1ccea91635374116/master/w_727,c_limit/buford-paul-bocuse-2.jpg, https://media.newyorker.com/photos/5a6633cd1ccea91635374116/master/w_1454,c_limit/buford-paul-bocuse-2.jpg 2x" data-reactid="136"/><img alt="" src="https://media.newyorker.com/photos/5a6633cd1ccea91635374116/master/w_727,c_limit/buford-paul-bocuse-2.jpg" aria-hidden="false" data-reactid="137"/></picture><button class="Button__button___2vDCa Button__seeMore___czyKz  ArticleLedeImage__button___3yTRZ" data-reactid="138"><div class="Button__content___d_gRZ" data-reactid="139"><i class="Button__icon___x9PXG Button__left___27AnT " data-reactid="140"><svg width="18" height="18" viewBox="0 0 18 18" xmlns="http://www.w3.org/2000/svg"><path d="M2 14.6l5-5 1.5 1.3-5 5H7v2H0v-7h2v3.6zM16 3.4l-5 5-1.5-1.3 5-5H11V0h7v7h-2V3.4z" fill="#A2A2A2" fill-rule="evenodd" /></svg></i></div></button></div></div></div><figcaption class="ImageCaption__captionWrapper___2h5XI  ImageCaption__default___3TPB5" data-reactid="141"><span class="ImageCaption__caption___1EOQO ImageCaption__caption___1EOQO" data-reactid="142"><div data-reactid="143"><p data-reactid="144"><!-- react-text: 145 -->Paul Bocuse, who died on Saturday, at the age of ninety-one, had an undeniably infectious culinary charisma. He was what people want a grand French chef to be.<!-- /react-text --></p></div></span><small class="ImageCaption__credit___rg3mC " data-reactid="146">Photograph by Jeff Pachoud / AFP / Getty</small></figcaption></figure></div></div></div><div id="articleBody" data-template="two-column" class="ArticleBody__articleBody___1GSGP" data-reactid="147"><div data-reactid="148"><div data-reactid="149"><div class="SectionBreak__sectionBreak___1ppA7" data-reactid="150"></div><div class="SectionBreak__sectionBreak___1ppA7" data-reactid="151"><p data-reactid="152"><!-- react-text: 153 -->Until <!-- /react-text --><span data-page="page_1" data-reactid="154"></span><!-- react-text: 155 -->1985 or thereabouts, Paul Bocuse was a very famous chef. He was a
household name, with an image so widely disseminated that just about
everyone could tell you what he looked like—the big lips, the toque, and
holding a live <!-- /react-text --><em data-reactid="156"><!-- react-text: 157 -->poulet de Bresse<!-- /react-text --></em><!-- react-text: 158 --> in his arms, say, and petting its
head—even if few people had any idea what his food might be like. Then,
around 1985 or so, Bocuse turned into an icon. One moment: famous guy at
stove. Then: Pope of restaurant people. He became the undisputed
emissary of the kitchen mission. He became Frenchness.<!-- /react-text --></p><p data-reactid="159"><!-- react-text: 160 -->It is not entirely clear how this happened, because (and Bocuse would
have agreed) there were, and are, more talented and savvy chefs around.
He never hosted a food show. Although a master of the photo op (hamming
it up in drag, stripping down to show off his tattoo of a Bresse
chicken, straddling his Harley Davidson), he rarely appeared on
television. He published cookbooks. None was a game-changer. Apart from
two forays abroad—one in Japan (which failed) and another at Walt Disney
World, in Florida (which thrives, run by his son Jérôme)—he never really
franchised his name. And yet, in the mysterious, ineffable way of these
things, Bocuse, who died on Saturday, at the age of ninety-one, had
something that no other chef has in the same abundance: an undeniably
infectious culinary charisma. Bocuse is what people want a grand French
chef to be.<!-- /react-text --></p><p data-reactid="161"><!-- react-text: 162 -->In 2009, I moved to Lyon with my family, to train to be a French chef
(not the “grand” kind, but more like a garden variety kitchen Frenchie).
We stayed five years. Bocuse was everywhere and nowhere. I enrolled in a
cooking school—<!-- /react-text --><em data-reactid="163"><!-- react-text: 164 -->the<!-- /react-text --></em><!-- react-text: 165 --> cooking school, l’Institut Paul Bocuse—where just
about every day there was a rumor that “Monsieur Paul” (as he is
referred to locally) was going to make an appearance. He never did,
although he regularly interrupted the instructors, calling them on their
cell phones during class. I was given hundreds of recipes and expected
to memorize them; they were mainly <!-- /react-text --><em data-reactid="166"><!-- react-text: 167 -->his<!-- /react-text --></em><!-- react-text: 168 --> recipes, I would eventually
learn. Around the city there were three murals devoted to his image,
including one that was directly across the street from our apartment on
the Saone river: he was what I looked at drinking my morning coffee.
Bridges were named after him. Les Halles, the indoor market, was
officially Les Halles Paul Bocuse. And so, too, was the world’s biggest
and most hysterical cooking competition, the biennial Bocuse d’Or.<!-- /react-text --></p><p data-reactid="169"><!-- react-text: 170 -->The event is organized like a World Cup (forty-eight nations competing,
each represented by a team of two, a chef and a deputy) and run like a
dog show, with dishes paraded through a stadium in front of twenty-six
judges. It culminates in an awards ceremony that manages to use the
iconography of the Olympics (with a podium for gold, silver, and
bronze), the Oscars (the “Bocuse d’Or” is a gold statuette in the image
of Paul Bocuse), and a New York City bar mitzvah (strobe lights, loud
oom-pah-pah music, and a ceiling drop of gold confetti). It is “tacky”
meets “technique,” but, despite the tackiness, the “technique” is real:
on display is the flashiest, most accomplished food on the planet.<!-- /react-text --></p><p data-reactid="171"><!-- react-text: 172 -->I attended my first Bocuse d’Or during our first year in France. I
showed up at 9 <!-- /react-text --><em class="small" data-reactid="173"><!-- react-text: 174 -->A.M.<!-- /react-text --></em><!-- react-text: 175 --> on the first day of what would be a two-day competition. The chefs had been there since five. Each two-member
team, plus a local <!-- /react-text --><em data-reactid="176"><!-- react-text: 177 -->commis<!-- /react-text --></em><!-- react-text: 178 -->, a slave-helper provided by l’Institut Paul
Bocuse, was crushed bumpingly into a mini-kitchen cubicle the size of a
changing room at a bad beach resort. The cubicles, side by side, were
spread out in a row, from left to right. They represented the “event.”
It was more rodeo arena than theatre. Facing the action were bleachers.
The mood in the mini-kitchen cubicles was of adrenaline and stress and
sweat. Every cook was intensely aware of a clock ticking. They were
focussed and very quiet. The bleachers were very not. There were
thousands of seats, and they were already filled. Many spectators
were in regional costume (the Japanese as samurais, the Mexicans in
sombreros) or draped in national flags (the Swedes, the Americans).
There was a very loud mariachi band, a very loud drum-and-bugle corps,
a very loud pit percussion team, some guys banging cymbals, and many
morons with stadium air horns. There wasn’t anything obvious to be
thundering about: no eviscerating of live goats, no fist pumping, no
chef standing up and saying, “Bam!” But the cheering and the chanting
never stopped.<!-- /react-text --></p><p data-reactid="179"><!-- react-text: 180 -->I felt uninformed and naïve—how could I have not known that this kind of
cooking (hunched-over, tortured people manipulating little things in
little ways) was a nationalist sport?—when, lo and behold, Bocuse
appeared.<!-- /react-text --></p><p data-reactid="181"><!-- react-text: 182 -->He had turned up backstage and taken it upon himself—the toque, the
whites, in a collar that looked like a French flag—to walk across the
floor. He padded by me, softly and stiffly, imposingly tall in his chef
posture, doing small papal waves, and seemed not to notice that a queue
of competitors was forming behind him. They were Japanese, Korean,
Indian, Brazilian, Norwegian . . . actually, I have no idea where they
were from, except that there were lots of them, and they seemed not to
be speaking either French or English. Only a moment before, they were
locked down<!-- /react-text --><span data-page="page_2" data-reactid="183"></span><!-- react-text: 184 --> in their panicky prep, when one of them noticed that Bocuse
was in the house and abandoned his station. He seemed to then realize, Hey,
it’s O.K. to follow the great man around—or at least he registered that
no one was making him stop—and called to the others in their
mini-cubicle kitchens, and urged them to join in. The line quickly
became much longer, conga-style.<!-- /react-text --></p><p data-reactid="185"><!-- react-text: 186 -->No one really knew what to do next—you can only get away with doing the
conga behind Bocuse for so long—when someone touched the great man
and dropped out of the queue, satisfied by the contact. The next person
in the queue touched a sleeve. Then it was a shoulder, the back of
Bocuse’s hand. An Asian chef was next up, and he seized Bocuse’s apron
by the hem, released it, and held the hand that had done the seizing by
its wrist, staring at it, shaking it, and screamed with wide-eyed
insanity as though his skin were burning. A cook fell to his knees, and
kissed the ground where the chef had tread (which, I don’t know, call me
prudish, seemed a little excessive).<!-- /react-text --></p><p data-reactid="187"><!-- react-text: 188 -->It got grabby. The line lost its linear definition, and, just as it was
seeming outright dangerous, Bocuse was gone. Fearing a mob moment,
handlers had emerged and ushered him through a backstage door.<!-- /react-text --></p><p data-reactid="189"><!-- react-text: 190 -->For the rest of the year, it became my mission to meet the great man
properly. I was told that I could reach him through his people. A wife,
a daughter, a son. I wrote them. I phoned. It was the daughter who
finally got back to me, and there was disdain in her reply: How dare you
think that you, a lowly writer guy, would have anything to say to
Monsieur Bocuse that is even remotely worth his time?<!-- /react-text --></p><p data-reactid="191"><!-- react-text: 192 -->Then I got lucky. I happened to be describing my frustrations in the bar
of La Mère Brazier, which is among the restaurants where Bocuse
trained. Mathieu Viannay is the chef, and he was standing nearby. What I
didn’t know is that Viannay was also a member of an “inner circle.” Its
members—basically, kick-ass chefs who understand the soul and suffering
of Lyonnais cooking—were awarded permanent access to the great man.<!-- /react-text --></p><p data-reactid="193"><!-- react-text: 194 -->Viannay interrupted my conversation. “You want to meet Monsieur Paul?”<!-- /react-text --></p><p data-reactid="195"><!-- react-text: 196 -->“Yes,” I said, “very much.”<!-- /react-text --></p><p data-reactid="197"><!-- react-text: 198 -->“Come by tomorrow morning at seven. Do not be late.”<!-- /react-text --></p><p data-reactid="199"><!-- react-text: 200 -->Viannay was waiting for me in his car. We crossed the Rhône River and
into the Third Arrondissement and in the direction of Les Halles de Lyon
Paul Bocuse. I then spotted the great chef’s vehicle, a massive black
Jeep Wrangler (American), with all the doo-dahs, spotlights on the roof,
a winch on the front bumper, and wide off-road tires (Michelin, of
course). It was parked slapdash across a sidewalk, in front of the
market-hall entrance. The sight confirmed two Lyonnais rumors: one, that
Bocuse always had his early-morning coffee at Les Halles; and, two, that
the police knew his vehicle and never ticketed it.<!-- /react-text --></p><p data-reactid="201"><!-- react-text: 202 -->For our part, we used a conventional parking lot, paid for it, entered
the hall, and stopped by a shellfish vendor named Leon, who gave us a
platter of oysters. (“Leon is the one you buy your oysters from,”
Viannay told me. “<!-- /react-text --><em data-reactid="203"><!-- react-text: 204 -->D’accord?<!-- /react-text --></em><!-- react-text: 205 -->”)<!-- /react-text --></p><p data-reactid="206"><!-- react-text: 207 -->We made our way to Le Boulanger, an informal café opposite the stand of
La Mère Richard, the city’s famous cheese lady, where Viannay paused to
point out the quality of what was on display. (“This is where you will
buy your cheese. <!-- /react-text --><em data-reactid="208"><!-- react-text: 209 -->D’accord<!-- /react-text --></em><!-- react-text: 210 -->?”)<!-- /react-text --></p><p data-reactid="211"><!-- react-text: 212 -->I looked across the aisle. There he was: Paul Bocuse, sitting by
himself, finishing a coffee.<!-- /react-text --></p><p data-reactid="213"><!-- react-text: 214 -->He was slumped slightly in his chair, and, wearing a black Pringle
cotton shirt, a worker’s cotton jacket, black trousers, and sneakers, he
looked like a regional train conductor at the end of his shift. On
spotting us, he stood up. He had shrunk considerably since I’d seen him
at the Bocuse d’Or. Then again, that last time, he had been in a
towering toque and heavy-heeled clogs, and doing that erect-chef-posture
routine. Without the gear, he was, frankly, a little naked. He seemed,
to me at least—and it is startling for me even to utter the thought—to
be almost normal. He was a man.<!-- /react-text --></p><p data-reactid="215"><!-- react-text: 216 -->There was a lot about him that I didn’t know, I realized. Actually, on
reflection, I knew nothing. Or, at least now—alone, suddenly, with the
deity himself—I was feeling that I knew nothing. I hadn’t yet eaten at
his main restaurant, L’Auberge—which has been awarded three Michelin
stars every year since 1965, making it the longest-running three-star
establishment in the <!-- /react-text --><em data-reactid="217"><!-- react-text: 218 -->Guide Michelin’s<!-- /react-text --></em><!-- react-text: 219 --> history—and was now angry with
myself that I hadn’t. I felt foolish, and with the great man shaking my
hand—among the greatest meetings in my life—I went instantly and
uncharacteristically shy. He told me to follow. “Viens,” he said gently.<!-- /react-text --></p><p data-reactid="220"><!-- react-text: 221 -->“<!-- /react-text --><em data-reactid="222"><!-- react-text: 223 -->Merci, chef<!-- /react-text --></em><!-- react-text: 224 -->,” I said weakly.<!-- /react-text --></p><p data-reactid="225"><!-- react-text: 226 -->He wanted to give me a tour.<!-- /react-text --></p><p data-reactid="227"><!-- react-text: 228 -->“<!-- /react-text --><em data-reactid="229"><!-- react-text: 230 -->Viens,”<!-- /react-text --></em><!-- react-text: 231 --> he repeated.<!-- /react-text --></p><p data-reactid="232"><!-- react-text: 233 -->“<!-- /react-text --><em data-reactid="234"><!-- react-text: 235 -->Merci, chef<!-- /react-text --></em><!-- react-text: 236 -->.”<!-- /react-text --></p><p data-reactid="237"><!-- react-text: 238 -->Since Bocuse knew everyone at Les Halles, and since it was an early hour
and very few shoppers had shown up yet, our slow passage through its
aisles was surprisingly intimate.<!-- /react-text --></p><p data-reactid="239"><!-- react-text: 240 -->“Here the charcuterie is very good,” he said, not unlike an owner of a
large estate showing off roses that were having a good year. “This is
chez Sibilia.”<!-- /react-text --></p><p data-reactid="241"><!-- react-text: 242 -->Viannay, standing nearby, whispered, “You will buy your<!-- /react-text --><span data-page="page_final" data-reactid="243"></span><!-- react-text: 244 --> charcuterie
nowhere else. <!-- /react-text --><em data-reactid="245"><!-- react-text: 246 -->D’accord<!-- /react-text --></em><!-- react-text: 247 -->?”<!-- /react-text --></p><p data-reactid="248"><!-- react-text: 249 -->“<!-- /react-text --><em data-reactid="250"><!-- react-text: 251 -->Oui, chef<!-- /react-text --></em><!-- react-text: 252 -->,” I whispered back.<!-- /react-text --></p><p data-reactid="253"><!-- react-text: 254 -->Bocuse and Sibilia, an imposing, no-nonsense woman, did kiss-kissy
bonjours and produced slices of rosette, the dried local sausage that
historians believe originated with mortadella and which the Lyonnais
crave whenever they leave town. They both stood around me while I tasted
it, studying my mouth, waiting for a verdict, as if such a tasting were
a matter of great importance. It was a shtick, of course, their journo
P.R. routine. I knew that. They knew that I knew that. But, meanwhile,
the overwhelming fact was this: Bocuse, <!-- /react-text --><em data-reactid="255"><!-- react-text: 256 -->the<!-- /react-text --></em><!-- react-text: 257 --> Paul Bocuse, was
hand-feeding me a <!-- /react-text --><em data-reactid="258"><!-- react-text: 259 -->saucisson<!-- /react-text --></em><!-- react-text: 260 -->, and I was remarkably O.K. with that.<!-- /react-text --></p><p data-reactid="261"><!-- react-text: 262 -->“<!-- /react-text --><em data-reactid="263"><!-- react-text: 264 -->Merci, chef<!-- /react-text --></em><!-- react-text: 265 -->.”<!-- /react-text --></p><p data-reactid="266"><!-- react-text: 267 -->Across the aisle was Les Volailles Clugnet, a bird vender. The
proprietor (“Bonjour, Pierre”), on seeing Bocuse, handed over one of his white
birds, unsolicited. Bocuse seized it between his two very large
hands and slapped it around, firmly but with affection, as though
handling the puppy of a favorite hunting dog.<!-- /react-text --></p><p data-reactid="268"><!-- react-text: 269 -->“The best chicken is the white breed from Bresse,” Bocuse said. “<!-- /react-text --><em data-reactid="270"><!-- react-text: 271 -->Tout
le monde sait ça.<!-- /react-text --></em><!-- react-text: 272 -->” Everyone knows this.<!-- /react-text --></p><p data-reactid="273"><!-- react-text: 274 -->“<!-- /react-text --><em data-reactid="275"><!-- react-text: 276 -->Merci, chef<!-- /react-text --></em><!-- react-text: 277 -->.”<!-- /react-text --></p><p data-reactid="278"><!-- react-text: 279 -->He handed the bird to me. It was heavy. Full. (Unlike the United States,
France doesn’t gut birds until they are about to be cooked.) I held on
to the chicken and looked at it with polite curiosity. I didn’t feel
confident enough to slap it around. Bocuse then did a riff on its
features, the red waddle draped across my thumb, the white feathers, the
dangling blue feet, and the many qualities that make the <!-- /react-text --><em data-reactid="280"><!-- react-text: 281 -->poulet de
Bresse<!-- /react-text --></em><!-- react-text: 282 --> a rare and scrumptious creature.<!-- /react-text --></p><p data-reactid="283"><!-- react-text: 284 -->“<!-- /react-text --><em data-reactid="285"><!-- react-text: 286 -->Merci, chef<!-- /react-text --></em><!-- react-text: 287 -->.”<!-- /react-text --></p><p data-reactid="288"><!-- react-text: 289 -->Viannay whispered, “Les Volailles Clugnet. Do you understand? The only
place.”<!-- /react-text --></p><p data-reactid="290"><!-- react-text: 291 -->“<!-- /react-text --><em data-reactid="292"><!-- react-text: 293 -->Oui, chef<!-- /react-text --></em><!-- react-text: 294 -->,” I whispered back.<!-- /react-text --></p><p data-reactid="295"><!-- react-text: 296 -->Then, like that, bang: Bocuse said goodbye.<!-- /react-text --></p><p data-reactid="297"><!-- react-text: 298 -->Leaving already? It was—I can’t deny it—a moment of shocking
disappointment. So soon? After such a long wait? I hadn’t even asked a
question.<!-- /react-text --></p><p data-reactid="299"><!-- react-text: 300 -->“<!-- /react-text --><em data-reactid="301"><!-- react-text: 302 -->À la prochaine fois<!-- /react-text --></em><!-- react-text: 303 -->,” he said.<!-- /react-text --></p><p data-reactid="304"><!-- react-text: 305 -->“Merci, chef,” I said sadly, but then brightened. <!-- /react-text --><em data-reactid="306"><!-- react-text: 307 -->À la prochaine
fois<!-- /react-text --></em><!-- react-text: 308 -->—until next time? Was he telling me that he’d be seeing me again?<!-- /react-text --></p></div></div></div></div><footer class="ArticleFooter__footer___3-wlJ" data-reactid="309"><!-- react-empty: 310 --><div class="ArticleContributors__bio___3XQjk" data-reactid="311"><div class="ArticleContributors__contributorWrapper___1CrIJ" data-reactid="312"><ul class="ArticleContributors__contributorBios___3_jrJ ArticleContributors__noAvatar___26vfu" data-reactid="313"><li data-reactid="314"><p class="ArticleContributors__contributorBioText___3m1QB" data-reactid="315">Bill Buford’s next book is “Dirt (Five Years in the Ancient Kitchens of Lyon, Where My Children Became Frenchies and I Trained to be a Chef),” which will be published in September. A former fiction editor at <em>The New Yorker</em>, he is also the author of “Among the Thugs” and “Heat.”</p><a class="Link__link___3dWao ArticleContributors__contributorBioLink___3ifgZ " href="/contributors/bill-buford" data-reactid="316"><!-- react-text: 317 -->Read more »<!-- /react-text --></a></li></ul></div></div><ul class="Tags__tags___1WDCj" data-reactid="318"><span class="Tags__heading___3sE5b" data-reactid="319">More:</span><li class="Tags__item___3xh9I" data-reactid="320"><a class="Link__link___3dWao  " href="/tag/food" data-reactid="321"><!-- react-text: 322 -->Food<!-- /react-text --></a></li><li class="Tags__item___3xh9I" data-reactid="323"><a class="Link__link___3dWao  " href="/tag/postscript" data-reactid="324"><!-- react-text: 325 -->Postscript<!-- /react-text --></a></li><li class="Tags__item___3xh9I" data-reactid="326"><a class="Link__link___3dWao  " href="/tag/cooking" data-reactid="327"><!-- react-text: 328 -->Cooking<!-- /react-text --></a></li><li class="Tags__item___3xh9I" data-reactid="329"><a class="Link__link___3dWao  " href="/tag/france" data-reactid="330"><!-- react-text: 331 -->France<!-- /react-text --></a></li></ul><!-- react-empty: 332 --><div class="" data-reactid="333"><form data-reactid="334"><section class="NewsletterSubscribe__newsletterComponent___1ZI2a" data-reactid="335"><!-- react-text: 336 --><!-- /react-text --><p class="NewsletterSubscribe__newsletterDek___29WDu" data-reactid="337">Get a recap of the week in culture every weekend, in your in-box. Sign up for our Culture Review newsletter.</p><div class="Input__inputContainer___3ZRfk [object Object]" data-reactid="338"><input type="email" autocomplete="on" class="Input__input___2fcvD " placeholder="Enter your e-mail" data-reactid="339"/><button class="Input__inputButton___2--U_" type="submit" data-reactid="340">Go</button></div><!-- react-text: 341 --><!-- /react-text --></section></form></div></footer></div></main><aside class="Layout__sidebar___2GcEE" data-reactid="342"><div class="ArticlePageSidebar__rail___PXqb4" data-reactid="343"><div class="ArticlePageSidebar__railItemContainer___3VZqQ" data-reactid="344"><div class="ArticlePageSidebar__railItem___1K-hs" data-reactid="345"><!-- react-empty: 346 --></div></div><div class="ArticlePageSidebar__railItemContainer___3VZqQ" data-reactid="347"><div class="ArticlePageSidebar__railItem___1K-hs" data-reactid="348"><div class="RecirculationMostPopular__default___3LVxh" data-reactid="349"><h3 class="RecirculationMostPopular__title___zwZ4R" data-reactid="350">Most Popular</h3><ul data-track-type="recirc" data-track-location="mod-most-popular" data-service="parsely" data-source="parsely" class="RecirculationMostPopular__itemsList___2lXkt" data-reactid="351"><li class="RecirculationMostPopular__item___2z9PN" data-track-type="recircItem" data-track-location="mod-most-popular" data-source="parsely" data-track-position="0" data-reactid="352"><span class="RecirculationMostPopular__counter___3YFww" data-reactid="353">1.</span><div class="RecirculationMostPopular__content___1no0_" data-reactid="354"><span class="RecirculationMostPopular__rubric___29E94" data-reactid="355">The Political Scene</span><a class="Link__link___3dWao  " href="/magazine/2018/01/29/jared-kushner-is-chinas-trump-card" data-reactid="356"><h5 class="RecirculationMostPopular__hed___1Yugq" data-reactid="357">Jared Kushner Is China’s Trump Card</h5></a><div class="Byline__theLatest___2viI5 " data-reactid="358"><p class="Byline__by___37lv8" data-reactid="359"><!-- react-text: 360 -->By <!-- /react-text --><a class="Link__link___3dWao  " title="Adam Entous" rel="author" href="/contributors/adam-entous" data-reactid="361"><!-- react-text: 362 -->Adam Entous<!-- /react-text --></a><span data-reactid="363"> and </span><a class="Link__link___3dWao  " title="Evan Osnos" rel="author" href="/contributors/evan-osnos" data-reactid="364"><!-- react-text: 365 -->Evan Osnos<!-- /react-text --></a></p><!-- react-text: 366 --><!-- /react-text --></div></div></li><li class="RecirculationMostPopular__item___2z9PN" data-track-type="recircItem" data-track-location="mod-most-popular" data-source="parsely" data-track-position="1" data-reactid="367"><span class="RecirculationMostPopular__counter___3YFww" data-reactid="368">2.</span><div class="RecirculationMostPopular__content___1no0_" data-reactid="369"><span class="RecirculationMostPopular__rubric___29E94" data-reactid="370">American Archives</span><a class="Link__link___3dWao  " href="/magazine/2018/01/29/the-lost-giant-of-american-literature" data-reactid="371"><h5 class="RecirculationMostPopular__hed___1Yugq" data-reactid="372">William Melvin Kelley: The Lost Giant of American Literature</h5></a><div class="Byline__theLatest___2viI5 " data-reactid="373"><p class="Byline__by___37lv8" data-reactid="374"><!-- react-text: 375 -->By <!-- /react-text --><a class="Link__link___3dWao  " title="Kathryn Schulz" rel="author" href="/contributors/kathryn-schulz" data-reactid="376"><!-- react-text: 377 -->Kathryn Schulz<!-- /react-text --></a></p><!-- react-text: 378 --><!-- /react-text --></div></div></li><li class="RecirculationMostPopular__item___2z9PN" data-track-type="recircItem" data-track-location="mod-most-popular" data-source="parsely" data-track-position="2" data-reactid="379"><span class="RecirculationMostPopular__counter___3YFww" data-reactid="380">3.</span><div class="RecirculationMostPopular__content___1no0_" data-reactid="381"><span class="RecirculationMostPopular__rubric___29E94" data-reactid="382">News Desk</span><a class="Link__link___3dWao  " href="/news/news-desk/did-schumer-cave" data-reactid="383"><h5 class="RecirculationMostPopular__hed___1Yugq" data-reactid="384">Did Schumer Cave on the Shutdown?</h5></a><div class="Byline__theLatest___2viI5 " data-reactid="385"><p class="Byline__by___37lv8" data-reactid="386"><!-- react-text: 387 -->By <!-- /react-text --><a class="Link__link___3dWao  " title="Benjamin Wallace-Wells" rel="author" href="/contributors/benjamin-wallace-wells" data-reactid="388"><!-- react-text: 389 -->Benjamin Wallace-Wells<!-- /react-text --></a></p><!-- react-text: 390 --><!-- /react-text --></div></div></li><li class="RecirculationMostPopular__item___2z9PN" data-track-type="recircItem" data-track-location="mod-most-popular" data-source="parsely" data-track-position="3" data-reactid="391"><span class="RecirculationMostPopular__counter___3YFww" data-reactid="392">4.</span><div class="RecirculationMostPopular__content___1no0_" data-reactid="393"><span class="RecirculationMostPopular__rubric___29E94" data-reactid="394">A Reporter at Large</span><a class="Link__link___3dWao  " href="/magazine/2018/01/29/a-prison-film-made-in-prison" data-reactid="395"><h5 class="RecirculationMostPopular__hed___1Yugq" data-reactid="396">A Prison Film Made in Prison</h5></a><div class="Byline__theLatest___2viI5 " data-reactid="397"><p class="Byline__by___37lv8" data-reactid="398"><!-- react-text: 399 -->By <!-- /react-text --><a class="Link__link___3dWao  " title="Nick Paumgarten" rel="author" href="/contributors/nick-paumgarten" data-reactid="400"><!-- react-text: 401 -->Nick Paumgarten<!-- /react-text --></a></p><!-- react-text: 402 --><!-- /react-text --></div></div></li><li class="RecirculationMostPopular__item___2z9PN" data-track-type="recircItem" data-track-location="mod-most-popular" data-source="parsely" data-track-position="4" data-reactid="403"><span class="RecirculationMostPopular__counter___3YFww" data-reactid="404">5.</span><div class="RecirculationMostPopular__content___1no0_" data-reactid="405"><span class="RecirculationMostPopular__rubric___29E94" data-reactid="406">Sporting Scene</span><a class="Link__link___3dWao  " href="/news/sporting-scene/the-patriots-marvelous-comeback-against-the-jaguars-and-the-muddled-joy-of-rooting-for-the-bad-guys" data-reactid="407"><h5 class="RecirculationMostPopular__hed___1Yugq" data-reactid="408">The Patriots’ Marvellous Comeback Against the Jaguars, and the Muddled Joy of Rooting for the Bad Guys</h5></a><div class="Byline__theLatest___2viI5 " data-reactid="409"><p class="Byline__by___37lv8" data-reactid="410"><!-- react-text: 411 -->By <!-- /react-text --><a class="Link__link___3dWao  " title="Ian Crouch" rel="author" href="/contributors/ian-crouch" data-reactid="412"><!-- react-text: 413 -->Ian Crouch<!-- /react-text --></a></p><!-- react-text: 414 --><!-- /react-text --></div></div></li></ul><a class="Link__link___3dWao RecirculationMostPopular__more___q6L9B RecirculationMostPopular__more___q6L9B" href="/popular?intcid=most-popular" data-reactid="415"><span data-reactid="416">our thirty most popular</span></a></div><!-- react-empty: 417 --></div></div><div class="ArticlePageSidebar__railItemContainer___3VZqQ" data-reactid="418"><div class="ArticlePageSidebar__railItem___1K-hs ArticlePageSidebar__last___2Ciro" data-reactid="419"><!-- react-empty: 420 --></div></div></div></aside><div class="Layout__social___jaQeD" data-reactid="421"><ul class="ArticleSocial__socialShare___2Skji ArticleSocial__sticky___2cA78" data-reactid="422"><li data-track-type="socialShare" data-track-location="default" data-source="facebook" class="SocialShareItem__icon___3QQZ0 SocialShareItem__facebook___15Vks " data-reactid="423"><a class="Link__link___3dWao  " href="https://www.facebook.com/sharer/sharer.php?u=https%3A%2F%2Fwww.newyorker.com%2Fculture%2Fannals-of-gastronomy%2Fthe-first-time-i-met-paul-bocuse&amp;display=popup&amp;ref=plugin" data-reactid="424"><div class="Link__content___6CYL1" data-reactid="425"><i class="Link__icon___3s7us Link__left___1QLUQ " data-reactid="426"><svg width="8" height="17" viewBox="0 0 8 17" xmlns="http://www.w3.org/2000/svg"><path d="M1.7323 17H5.2V8.499h2.3135L7.82 5.5698H5.2l.0037-1.4667c0-.7634.072-1.1733 1.154-1.1733h1.4458V0h-2.314C2.7104 0 1.7324 1.4214 1.7324 3.811v1.7588H0v2.9297h1.7323V17z" fill="#000" fill-rule="evenodd"/></svg></i></div></a></li><li data-track-type="socialShare" data-track-location="default" data-source="twitter" class="SocialShareItem__icon___3QQZ0 SocialShareItem__twitter___1KPYQ " data-reactid="427"><a class="Link__link___3dWao  " href="https://twitter.com/intent/tweet?original_referer=https%3A%2F%2Fwww.newyorker.com%2Fculture%2Fannals-of-gastronomy%2Fthe-first-time-i-met-paul-bocuse&amp;text=The%20First%20Time%20I%20Met%20Paul%20Bocuse&amp;tw_p=tweetbutton&amp;url=https%3A%2F%2Fwww.newyorker.com%2Fculture%2Fannals-of-gastronomy%2Fthe-first-time-i-met-paul-bocuse" data-reactid="428"><div class="Link__content___6CYL1" data-reactid="429"><i class="Link__icon___3s7us Link__left___1QLUQ " data-reactid="430"><svg width="19" height="15" viewBox="0 0 19 15" xmlns="http://www.w3.org/2000/svg"><path d="M18.223 1.739c-.654.295-1.3565.4943-2.094.584.7527-.459 1.3308-1.1858 1.603-2.0518-.7045.425-1.4847.7336-2.3152.9C14.7518.4504 13.8043 0 12.7558 0c-2.0137 0-3.646 1.6605-3.646 3.7084 0 .2908.032.5737.0943.8452-3.03-.1547-5.7164-1.631-7.5146-3.8748-.3138.5478-.4936 1.1848-.4936 1.8644 0 1.2866.6436 2.4217 1.622 3.0868-.5977-.0192-1.16-.186-1.6515-.464l-.0003.0468c0 1.7968 1.2567 3.2957 2.9246 3.6364-.306.0847-.628.13-.9605.13-.235 0-.4632-.0232-.686-.0665.464 1.4733 1.8105 2.5455 3.406 2.5755-1.248.9946-2.82 1.5875-4.528 1.5875-.2943 0-.5845-.0176-.8698-.0518 1.6136 1.052 3.53 1.666 5.5888 1.666 6.706 0 10.3733-5.651 10.3733-10.5515 0-.1608-.0036-.3208-.0106-.4798.7123-.5228 1.3304-1.176 1.8192-1.9197" fill="#000" fill-rule="evenodd"/></svg></i></div></a></li><li data-track-type="socialShare" data-track-location="default" data-source="email" class="SocialShareItem__icon___3QQZ0 SocialShareItem__email___aXa4V " data-reactid="431"><a class="Link__link___3dWao  " href="mailto:?subject=From%20newyorker.com:%20The%20First%20Time%20I%20Met%20Paul%20Bocuse&amp;body=Bocuse%2C%20who%20died%20on%20Saturday%2C%20at%20the%20age%20of%20ninety-one%2C%20had%20an%20undeniably%20infectious%20culinary%20charisma.%20He%20was%20what%20people%20want%20a%20grand%20French%0Achef%20to%20be.%20%0D%0Ahttps%3A%2F%2Fwww.newyorker.com%2Fculture%2Fannals-of-gastronomy%2Fthe-first-time-i-met-paul-bocuse" data-reactid="432"><div class="Link__content___6CYL1" data-reactid="433"><i class="Link__icon___3s7us Link__left___1QLUQ " data-reactid="434"><svg width="19" height="14" viewBox="0 0 19 14" xmlns="http://www.w3.org/2000/svg"><path d="M9.7902 9.4853l-.003-.003-.004.003-6.9808-7.071 1.396-1.414 5.5888 5.66 5.588-5.66 1.396 1.414-6.981 7.071zm-8.9624 4.515h17.7707v-14H.8278v14z" fill="#000" fill-rule="evenodd"/></svg></i></div></a></li><li data-track-type="socialShare" data-track-location="default" data-source="print" class="SocialShareItem__icon___3QQZ0 SocialShareItem__print___2zPf0 " data-reactid="435"><a class="Link__link___3dWao  " href="#" data-reactid="436"><div class="Link__content___6CYL1" data-reactid="437"><i class="Link__icon___3s7us Link__left___1QLUQ " data-reactid="438"><svg width="17" height="17" viewBox="0 0 17 17" xmlns="http://www.w3.org/2000/svg"><g fill="#FFF" fill-rule="evenodd"><path d="M4.9688 1.988h8.008v2.009h.999V1.988c0-.559-.456-.999-.999-.999h-8.008c-.559 0-.999.456-.999.999v2.009h.999V1.988zM14.9857 8.0037c-.559 0-.999-.456-.999-.999 0-.559.456-.999.999-.999.559 0 .999.456.999.999.016.543-.44.999-.999.999zm-1.993 8.008h-8.024v-7.009h8.008v7.015h.016v-.006zm3.008-11.016H1.9777c-.56 0-.999.456-.999.999v5.001c0 .559.456.998.999.998h2.008v4.007c0 .559.456.999.999.999h8.008c.56 0 .999-.456.999-.999v-3.985h2.009c.559 0 .999-.456.999-.999v-5.022c0-.543-.456-.999-.999-.999z"/><path d="M5.979 11.012h4.007v-.999H5.979M5.979 13.01h6.016v-.999H5.979M5.979 15.018h6.016v-.999H5.979"/></g></svg></i></div></a></li></ul></div></div><div class="FeaturedVideo__sectionColumn___2zRYf" data-reactid="439"><div class="FeaturedVideo__sectionRow___2bKsR" data-reactid="440"><h5 class="FeaturedVideo__sectionTitle___1wlGn" data-reactid="441">Video</h5><div class="Video__featuredVideo___3_2hs" data-reactid="442"><div class="Video__videoContainer___2gi7E" role="button" tabindex="0" data-reactid="443"><div class="component-lazy pending" data-component="Lazy" data-reactid="444"></div></div><p class="Video__title___3QAw2" data-reactid="445">A Priceless Pizzeria in Brooklyn</p><p class="Video__description___1pPO8" data-reactid="446">Domenico DeMarco has been making pizzas at Brooklyn’s famous Di Fara Pizza since 1965.</p></div></div></div><section class="RecirculationRecommendedStories__recommended___3KI8A" data-reactid="447"><div class="RecirculationRecommendedStories__column___B7XcW" data-track-type="recirc" data-track-location="mod-related" data-service="none, parsely" data-source="parsely" data-reactid="448"><h5 class="RecirculationRecommendedStories__header___y7h0R" data-reactid="449">Recommended Stories</h5><div class="RecirculationRecommendedStories__row___1MVU6" data-reactid="450"><div class="RecirculationRecommendedStories__story___1S5ZJ" data-track-type="recircItem" data-track-location="mod-related" data-service="tbd" data-source="curated" data-track-position="0" data-reactid="451"><a class="Link__link___3dWao RecirculationRecommendedStories__image___3M88v " href="/culture/annals-of-gastronomy/the-butter-chicken-lady-who-made-indian-cooks-love-the-instant-pot" data-reactid="452"><div class="Image__image___1PhYl" role="false" tabindex="false" data-reactid="453"><div class="component-lazy pending" data-component="Lazy" data-reactid="454"></div></div></a><a class="Link__link___3dWao Rubric__rubric___3hrQE RecirculationRecommendedStories__rubric___2phUL " href="/culture/annals-of-gastronomy" title="Annals of Gastronomy" data-reactid="455"><!-- react-text: 456 -->Annals of Gastronomy<!-- /react-text --></a><a class="Link__link___3dWao  " href="/culture/annals-of-gastronomy/the-butter-chicken-lady-who-made-indian-cooks-love-the-instant-pot" data-reactid="457"><h2 class="RecirculationRecommendedStories__hed___3cnVL" data-reactid="458">The “Butter-Chicken Lady” Who Made Indian Cooks Love the Instant Pot</h2></a><h2 class="RecirculationRecommendedStories__dek___1rygJ" data-reactid="459">Urvashi Pitre’s most hard-earned fans are her fellow-Indians, who tend to  be skeptical that a dish whipped up in fifteen minutes could qualify as authentic.</h2><p class="RecirculationRecommendedStories__byline___eynVy" data-reactid="460"><!-- react-text: 461 -->By <!-- /react-text --><a class="Link__link___3dWao  " href="/contributors/priya-krishna" title="Priya Krishna" rel="author" data-reactid="462"><!-- react-text: 463 -->Priya Krishna<!-- /react-text --></a></p><span class="RecirculationRecommendedStories__timestamp___1YE3P" data-reactid="464">Jan. 11, 2018</span></div><div class="RecirculationRecommendedStories__story___1S5ZJ" data-track-type="recircItem" data-track-location="mod-related" data-service="tbd" data-source="curated" data-track-position="1" data-reactid="465"><a class="Link__link___3dWao RecirculationRecommendedStories__image___3M88v " href="/culture/annals-of-gastronomy/in-praise-of-flour-tortillas-an-unsung-jewel-of-the-us-mexico-borderlands" data-reactid="466"><div class="Image__image___1PhYl" role="false" tabindex="false" data-reactid="467"><div class="component-lazy pending" data-component="Lazy" data-reactid="468"></div></div></a><a class="Link__link___3dWao Rubric__rubric___3hrQE RecirculationRecommendedStories__rubric___2phUL " href="/culture/annals-of-gastronomy" title="Annals of Gastronomy" data-reactid="469"><!-- react-text: 470 -->Annals of Gastronomy<!-- /react-text --></a><a class="Link__link___3dWao  " href="/culture/annals-of-gastronomy/in-praise-of-flour-tortillas-an-unsung-jewel-of-the-us-mexico-borderlands" data-reactid="471"><h2 class="RecirculationRecommendedStories__hed___3cnVL" data-reactid="472">In Praise of Flour Tortillas, an Unsung Jewel of the U.S.-Mexico Borderlands</h2></a><h2 class="RecirculationRecommendedStories__dek___1rygJ" data-reactid="473">Recent Mexican immigrants deride them as a gringo quirk. Foodie purists dismiss them as not “real” Mexican food. But they can be revelatory.</h2><p class="RecirculationRecommendedStories__byline___eynVy" data-reactid="474"><!-- react-text: 475 -->By <!-- /react-text --><a class="Link__link___3dWao  " href="/contributors/gustavo-arellano" title="Gustavo Arellano" rel="author" data-reactid="476"><!-- react-text: 477 -->Gustavo Arellano<!-- /react-text --></a></p><span class="RecirculationRecommendedStories__timestamp___1YE3P" data-reactid="478">Jan. 13, 2018</span></div></div><div class="RecirculationRecommendedStories__row___1MVU6" data-reactid="479"><div class="RecirculationRecommendedStories__story___1S5ZJ" data-track-type="recircItem" data-track-location="mod-related" data-service="tbd" data-source="parsely" data-track-position="0" data-reactid="480"><a class="Link__link___3dWao RecirculationRecommendedStories__image___3M88v " href="/magazine/2018/01/29/jared-kushner-is-chinas-trump-card" data-reactid="481"><div class="Image__image___1PhYl" role="false" tabindex="false" data-reactid="482"><div class="component-lazy pending" data-component="Lazy" data-reactid="483"></div></div></a><span class="Rubric__rubric___3hrQE RecirculationRecommendedStories__rubric___2phUL" data-reactid="484">The Political Scene</span><a class="Link__link___3dWao  " href="/magazine/2018/01/29/jared-kushner-is-chinas-trump-card" data-reactid="485"><h2 class="RecirculationRecommendedStories__hed___3cnVL" data-reactid="486">Jared Kushner Is China’s Trump Card</h2></a><h2 class="RecirculationRecommendedStories__dek___1rygJ" data-reactid="487">China’s suspect courtship with Jared Kushner.</h2><p class="RecirculationRecommendedStories__byline___eynVy" data-reactid="488"><!-- react-text: 489 -->By <!-- /react-text --><a class="Link__link___3dWao  " href="/contributors/adam-entous" title="Adam Entous" rel="author" data-reactid="490"><!-- react-text: 491 -->Adam Entous<!-- /react-text --></a></p><span class="RecirculationRecommendedStories__timestamp___1YE3P" data-reactid="492"></span></div><div class="RecirculationRecommendedStories__story___1S5ZJ" data-track-type="recircItem" data-track-location="mod-related" data-service="tbd" data-source="parsely" data-track-position="1" data-reactid="493"><a class="Link__link___3dWao RecirculationRecommendedStories__image___3M88v " href="/magazine/2018/01/29/the-lost-giant-of-american-literature" data-reactid="494"><div class="Image__image___1PhYl" role="false" tabindex="false" data-reactid="495"><div class="component-lazy pending" data-component="Lazy" data-reactid="496"></div></div></a><span class="Rubric__rubric___3hrQE RecirculationRecommendedStories__rubric___2phUL" data-reactid="497">American Archives</span><a class="Link__link___3dWao  " href="/magazine/2018/01/29/the-lost-giant-of-american-literature" data-reactid="498"><h2 class="RecirculationRecommendedStories__hed___3cnVL" data-reactid="499">William Melvin Kelley: The Lost Giant of American Literature</h2></a><h2 class="RecirculationRecommendedStories__dek___1rygJ" data-reactid="500">A brilliant début launched the career of a singular talent. How did he get lost?</h2><p class="RecirculationRecommendedStories__byline___eynVy" data-reactid="501"><!-- react-text: 502 -->By <!-- /react-text --><a class="Link__link___3dWao  " href="/contributors/kathryn-schulz" title="Kathryn Schulz" rel="author" data-reactid="503"><!-- react-text: 504 -->Kathryn Schulz<!-- /react-text --></a></p><span class="RecirculationRecommendedStories__timestamp___1YE3P" data-reactid="505"></span></div></div></div></section></article><!-- react-empty: 506 --></section><footer data-track-type="nav" data-track-location="page-footer" class="SiteFooter__footer___2CFaR" data-reactid="507"><div class="FooterPromoLinks__promoLinks___3HDWB" data-reactid="508"><div class="FooterPromoLinks__container___fzxb9" data-reactid="509"><div class="FooterPromoLinks__featureLinks___17S0N" data-reactid="510"><div data-reactid="511"><h4 data-reactid="512"><a class="Link__link___3dWao  " href="/podcast/the-new-yorker-radio-hour" data-reactid="513"><!-- react-text: 514 -->Listen to the New Yorker Radio Hour<!-- /react-text --></a></h4><h4 data-reactid="515"><a class="Link__link___3dWao  " data-track-type="shop" data-track-location="magazine-cover" href="http://www.condenaststore.com/-st/New-Yorker-Covers-Prints_c147247_.htm" target="_blank" data-reactid="516"><!-- react-text: 517 -->Buy the Cover<!-- /react-text --></a></h4><h4 data-reactid="518"><a class="Link__link___3dWao  " href="/jigsaw" data-reactid="519"><!-- react-text: 520 -->Play the Jigsaw Puzzle<!-- /react-text --></a></h4></div></div></div></div><div class="SiteFooter__footerContainer___2Ueuo" data-reactid="521"><div class="SiteFooter__footerContent___36hmA" data-reactid="522"><div class="FooterDirectory__directoryContainer___1tqT-" data-reactid="523"><div class="FooterDirectory__directory___C4Dvz" data-reactid="524"><div class="FooterDirectory__panel___2uuX1" data-reactid="525"><h4 data-reactid="526"><a href="/news" data-reactid="527">News &amp; Politics</a></h4><ul data-reactid="528"><li data-reactid="529"><a class="Link__link___3dWao  " href="/news/daily-comment" data-reactid="530"><!-- react-text: 531 -->Daily Comment<!-- /react-text --></a></li><li data-reactid="532"><a class="Link__link___3dWao  " href="/news/john-cassidy" data-reactid="533"><!-- react-text: 534 -->John Cassidy<!-- /react-text --></a></li><li data-reactid="535"><a class="Link__link___3dWao  " href="/news/amy-davidson" data-reactid="536"><!-- react-text: 537 -->Amy Davidson<!-- /react-text --></a></li><li data-reactid="538"><a class="Link__link___3dWao  " href="/news/ryan-lizza" data-reactid="539"><!-- react-text: 540 -->Ryan Lizza<!-- /react-text --></a></li><li data-reactid="541"><a class="Link__link___3dWao  " href="/news/benjamin-wallace-wells" data-reactid="542"><!-- react-text: 543 -->Benjamin Wallace-Wells<!-- /react-text --></a></li></ul></div><div class="FooterDirectory__panel___2uuX1" data-reactid="544"><h4 data-reactid="545"><a href="/culture" data-reactid="546">Culture</a></h4><ul data-reactid="547"><li data-reactid="548"><a class="Link__link___3dWao  " href="/culture/cultural-comment" data-reactid="549"><!-- react-text: 550 -->Cultural Comment<!-- /react-text --></a></li><li data-reactid="551"><a class="Link__link___3dWao  " href="/culture/culture-desk" data-reactid="552"><!-- react-text: 553 -->Culture Desk<!-- /react-text --></a></li><li data-reactid="554"><a class="Link__link___3dWao  " href="/goings-on-about-town" data-reactid="555"><!-- react-text: 556 -->Goings On About Town<!-- /react-text --></a></li><li data-reactid="557"><a class="Link__link___3dWao  " href="/magazine/critics" data-reactid="558"><!-- react-text: 559 -->The Critics<!-- /react-text --></a></li><li data-reactid="560"><a class="Link__link___3dWao  " href="/culture/jia-tolentino" data-reactid="561"><!-- react-text: 562 -->Jia Tolentino<!-- /react-text --></a></li><li data-reactid="563"><a class="Link__link___3dWao  " href="/culture/persons-of-interest" data-reactid="564"><!-- react-text: 565 -->Persons of Interest<!-- /react-text --></a></li></ul></div><div class="FooterDirectory__panel___2uuX1" data-reactid="566"><h4 data-reactid="567"><a href="/business" data-reactid="568">Business, Science &amp; Tech</a></h4><ul data-reactid="569"><li data-reactid="570"><a class="Link__link___3dWao  " href="/business/currency" data-reactid="571"><!-- react-text: 572 -->Currency<!-- /react-text --></a></li><li data-reactid="573"><a class="Link__link___3dWao  " href="/tech/elements" data-reactid="574"><!-- react-text: 575 -->Elements<!-- /react-text --></a></li><li data-reactid="576"><a class="Link__link___3dWao  " href="/science/maria-konnikova" data-reactid="577"><!-- react-text: 578 -->Maria Konnikova<!-- /react-text --></a></li></ul></div><div class="FooterDirectory__panel___2uuX1" data-reactid="579"><h4 data-reactid="580"><a href="/humor" data-reactid="581">Humor</a></h4><ul data-reactid="582"><li data-reactid="583"><a class="Link__link___3dWao  " href="/humor/daily-shouts" data-reactid="584"><!-- react-text: 585 -->Daily Shouts<!-- /react-text --></a></li><li data-reactid="586"><a class="Link__link___3dWao  " href="/magazine/shouts-murmurs" data-reactid="587"><!-- react-text: 588 -->Shouts &amp; Murmurs<!-- /react-text --></a></li><li data-reactid="589"><a class="Link__link___3dWao  " href="/humor/borowitz-report" data-reactid="590"><!-- react-text: 591 -->The Borowitz Report<!-- /react-text --></a></li></ul></div><div class="FooterDirectory__panel___2uuX1" data-reactid="592"><h4 data-reactid="593"><a href="/cartoons" data-reactid="594">Cartoons</a></h4><ul data-reactid="595"><li data-reactid="596"><a class="Link__link___3dWao  " href="/cartoons/daily-cartoon" data-reactid="597"><!-- react-text: 598 -->Daily Cartoon<!-- /react-text --></a></li><li data-reactid="599"><a class="Link__link___3dWao  " href="http://contest.newyorker.com/" data-reactid="600"><!-- react-text: 601 -->Cartoon Caption Contest<!-- /react-text --></a></li><li data-reactid="602"><a class="Link__link___3dWao  " data-track-type="shop" data-track-location="cartoon-shop" href="https://condenaststore.com/conde-nast-brand/cartoonbank?utm_medium=referral" data-reactid="603"><!-- react-text: 604 -->Cartoon Bank<!-- /react-text --></a></li></ul></div><div class="FooterDirectory__panel___2uuX1" data-reactid="605"><h4 data-reactid="606"><a href="/books" data-reactid="607">Books &amp; Fiction</a></h4><ul data-reactid="608"><li data-reactid="609"><a class="Link__link___3dWao  " href="/books/page-turner" data-reactid="610"><!-- react-text: 611 -->Page-Turner<!-- /react-text --></a></li><li data-reactid="612"><a class="Link__link___3dWao  " href="/magazine/books" data-reactid="613"><!-- react-text: 614 -->Books<!-- /react-text --></a></li><li data-reactid="615"><a class="Link__link___3dWao  " href="/magazine/literary-lives" data-reactid="616"><!-- react-text: 617 -->Literary Lives<!-- /react-text --></a></li><li data-reactid="618"><a class="Link__link___3dWao  " href="/magazine/poems" data-reactid="619"><!-- react-text: 620 -->Poems<!-- /react-text --></a></li><li data-reactid="621"><a class="Link__link___3dWao  " href="/magazine/fiction" data-reactid="622"><!-- react-text: 623 -->Fiction<!-- /react-text --></a></li></ul></div><div class="FooterDirectory__panel___2uuX1" data-reactid="624"><h4 data-reactid="625"><a href="/magazine" data-reactid="626">Magazine</a></h4><ul data-reactid="627"><li data-reactid="628"><a class="Link__link___3dWao  " href="/magazine" data-reactid="629"><!-- react-text: 630 -->This Week&#x27;s Issue<!-- /react-text --></a></li><li data-reactid="631"><a class="Link__link___3dWao  " href="/archive" data-reactid="632"><!-- react-text: 633 -->Archive<!-- /react-text --></a></li><li data-reactid="634"><a class="Link__link___3dWao  " href="/subscribe" data-reactid="635"><!-- react-text: 636 -->Subscribe<!-- /react-text --></a></li></ul></div><div class="FooterDirectory__panel___2uuX1" data-reactid="637"><h4 data-reactid="638"><a href="/tag/photography" data-reactid="639">Photography</a></h4><ul data-reactid="640"><li data-reactid="641"><a class="Link__link___3dWao  " href="/culture/photo-booth" data-reactid="642"><!-- react-text: 643 -->Photo Booth<!-- /react-text --></a></li><li data-reactid="644"><a class="Link__link___3dWao  " href="/magazine/portfolio" data-reactid="645"><!-- react-text: 646 -->Portfolio<!-- /react-text --></a></li></ul></div><div class="FooterDirectory__panel___2uuX1" data-reactid="647"><h4 data-reactid="648"><a href="http://video.newyorker.com/" data-reactid="649">Video</a></h4><ul data-reactid="650"><li data-reactid="651"><a class="Link__link___3dWao  " href="http://video.newyorker.com/genres/Culture" data-reactid="652"><!-- react-text: 653 -->Culture<!-- /react-text --></a></li><li data-reactid="654"><a class="Link__link___3dWao  " href="http://video.newyorker.com/genres/Humor" data-reactid="655"><!-- react-text: 656 -->Humor<!-- /react-text --></a></li><li data-reactid="657"><a class="Link__link___3dWao  " href="http://video.newyorker.com/genres/Politics" data-reactid="658"><!-- react-text: 659 -->Politics<!-- /react-text --></a></li><li data-reactid="660"><a class="Link__link___3dWao  " href="http://video.newyorker.com/genres/Business" data-reactid="661"><!-- react-text: 662 -->Business<!-- /react-text --></a></li><li data-reactid="663"><a class="Link__link___3dWao  " href="http://video.newyorker.com/genres/Science%20&amp;%20Tech" data-reactid="664"><!-- react-text: 665 -->Science &amp; Tech<!-- /react-text --></a></li><li data-reactid="666"><a class="Link__link___3dWao  " href="http://video.newyorker.com/genres/Books" data-reactid="667"><!-- react-text: 668 -->Books<!-- /react-text --></a></li><li data-reactid="669"><a class="Link__link___3dWao  " href="http://video.newyorker.com/genres/Sports" data-reactid="670"><!-- react-text: 671 -->Sports<!-- /react-text --></a></li></ul></div><div class="FooterDirectory__panel___2uuX1" data-reactid="672"><h4 data-reactid="673"><a href="/podcast" data-reactid="674">Podcasts</a></h4><ul data-reactid="675"><li data-reactid="676"><a class="Link__link___3dWao  " href="/podcast/the-new-yorker-radio-hour" data-reactid="677"><!-- react-text: 678 -->The New Yorker Radio Hour<!-- /react-text --></a></li><li data-reactid="679"><a class="Link__link___3dWao  " href="/podcast/political-scene" data-reactid="680"><!-- react-text: 681 -->Political Scene<!-- /react-text --></a></li><li data-reactid="682"><a class="Link__link___3dWao  " href="/podcast/the-authors-voice" data-reactid="683"><!-- react-text: 684 -->The Writer&#x27;s Voice<!-- /react-text --></a></li><li data-reactid="685"><a class="Link__link___3dWao  " href="/podcast/fiction" data-reactid="686"><!-- react-text: 687 -->Fiction<!-- /react-text --></a></li><li data-reactid="688"><a class="Link__link___3dWao  " href="/podcast/poetry" data-reactid="689"><!-- react-text: 690 -->Poetry<!-- /react-text --></a></li><li data-reactid="691"><a class="Link__link___3dWao  " href="/podcast/out-loud" data-reactid="692"><!-- react-text: 693 -->Out Loud<!-- /react-text --></a></li></ul></div><div class="FooterDirectory__panel___2uuX1" data-reactid="694"><h4 data-reactid="695">More</h4><ul data-reactid="696"><li data-reactid="697"><a class="Link__link___3dWao  " href="http://w1.buysub.com/servlet/CSGateway?cds_mag_code=NYR" target="_blank" data-reactid="698"><!-- react-text: 699 -->Customer Care<!-- /react-text --></a></li><li data-reactid="700"><a class="Link__link___3dWao  " data-track-type="shop" data-track-location="magazine-cover" href="http://www.condenaststore.com/-st/New-Yorker-Covers-Prints_c147247_.htm" target="_blank" data-reactid="701"><!-- react-text: 702 -->Buy the Cover<!-- /react-text --></a></li><li data-reactid="703"><a class="Link__link___3dWao  " href="/digital-editions" data-reactid="704"><!-- react-text: 705 -->Apps<!-- /react-text --></a></li><li data-reactid="706"><a class="Link__link___3dWao  " href="/jigsaw" data-reactid="707"><!-- react-text: 708 -->Jigsaw Puzzle<!-- /react-text --></a></li><li data-reactid="709"><a class="Link__link___3dWao  " href="https://projects.newyorker.com/securedrop/" data-reactid="710"><!-- react-text: 711 -->SecureDrop<!-- /react-text --></a></li><li data-reactid="712"><a class="Link__link___3dWao  " href="http://www.condenaststore.com/~/new-yorker-store" data-reactid="713"><!-- react-text: 714 -->Store<!-- /react-text --></a></li><li data-reactid="715"><a class="Link__link___3dWao  " href="/about/feeds" data-reactid="716"><!-- react-text: 717 -->RSS<!-- /react-text --></a></li><li data-reactid="718"><a class="Link__link___3dWao  " href="/sitemap" data-reactid="719"><!-- react-text: 720 -->Site Map<!-- /react-text --></a></li></ul></div><div class="FooterDirectory__panel___2uuX1" data-reactid="721"><h4 data-reactid="722"><a href="https://www.newyorker.com/newsletters" data-reactid="723">Newsletters</a></h4><ul data-reactid="724"><li data-reactid="725"><a class="Link__link___3dWao  " href="https://www.newyorker.com/newsletters/daily-newsletter" data-reactid="726"><!-- react-text: 727 -->The Daily<!-- /react-text --></a></li><li data-reactid="728"><a class="Link__link___3dWao  " href="https://www.newyorker.com/newsletters/culture-review-newsletter" data-reactid="729"><!-- react-text: 730 -->Culture Review<!-- /react-text --></a></li><li data-reactid="731"><a class="Link__link___3dWao  " href="https://www.newyorker.com/newsletters/podcasts-newsletter" data-reactid="732"><!-- react-text: 733 -->Podcasts<!-- /react-text --></a></li><li data-reactid="734"><a class="Link__link___3dWao  " href="https://www.newyorker.com/newsletters/cartoons-newsletter" data-reactid="735"><!-- react-text: 736 -->Cartoons<!-- /react-text --></a></li><li data-reactid="737"><a class="Link__link___3dWao  " href="https://www.newyorker.com/newsletters/john-cassidy-newsletter" data-reactid="738"><!-- react-text: 739 -->John Cassidy<!-- /react-text --></a></li><li data-reactid="740"><a class="Link__link___3dWao  " href="https://www.newyorker.com/newsletters/andy-borowitz-newsletter" data-reactid="741"><!-- react-text: 742 -->The Borowitz Report<!-- /react-text --></a></li><li data-reactid="743"><a class="Link__link___3dWao  " href="https://www.newyorker.com/newsletters/fiction-newsletter" data-reactid="744"><!-- react-text: 745 -->Fiction<!-- /react-text --></a></li><li data-reactid="746"><a class="Link__link___3dWao  " href="https://www.newyorker.com/newsletters/go-newsletter" data-reactid="747"><!-- react-text: 748 -->Goings On About Town<!-- /react-text --></a></li></ul></div><div class="FooterDirectory__panel___2uuX1" data-reactid="749"><h4 data-reactid="750"><a href="/about/us" data-reactid="751">About Us</a></h4><ul data-reactid="752"><li data-reactid="753"><a class="Link__link___3dWao  " href="/about/us" data-reactid="754"><!-- react-text: 755 -->About<!-- /react-text --></a></li><li data-reactid="756"><a class="Link__link___3dWao  " href="/about/careers" data-reactid="757"><!-- react-text: 758 -->Careers<!-- /react-text --></a></li><li data-reactid="759"><a class="Link__link___3dWao  " href="/about/contact" data-reactid="760"><!-- react-text: 761 -->Contact<!-- /react-text --></a></li><li data-reactid="762"><a class="Link__link___3dWao  " href="/about/faq" data-reactid="763"><!-- react-text: 764 -->FAQ<!-- /react-text --></a></li><li data-reactid="765"><a class="Link__link___3dWao  " href="http://www.condenast.com/brands/the-new-yorker" data-reactid="766"><!-- react-text: 767 -->Media Kit<!-- /react-text --></a></li><li data-reactid="768"><a class="Link__link___3dWao  " href="/about/press" data-reactid="769"><!-- react-text: 770 -->Press<!-- /react-text --></a></li><li data-reactid="771"><a class="Link__link___3dWao  " href="/about/accessibility-help/" data-reactid="772"><!-- react-text: 773 -->Accessibility Help<!-- /react-text --></a></li></ul></div></div></div><div class="SiteFooter__directoryDropdownContainer___2BuP0" data-reactid="774"><div class="SelectLinks__dropdown___24BA0" data-reactid="775"><select data-reactid="776"><option selected="" value="" disabled="" data-reactid="777">Sections</option><option value="/news" data-reactid="778">News & Politics</option><option value="/culture" data-reactid="779">Culture</option><option value="/business" data-reactid="780">Business, Science & Tech</option><option value="/humor" data-reactid="781">Humor</option><option value="/cartoons" data-reactid="782">Cartoons</option><option value="/books" data-reactid="783">Books & Fiction</option><option value="/magazine" data-reactid="784">Magazine</option><option value="/tag/photography" data-reactid="785">Photography</option><option value="http://video.newyorker.com/" data-reactid="786">Video</option><option value="/podcast" data-reactid="787">Podcasts</option><option selected="" data-reactid="788">More</option><option value="https://www.newyorker.com/newsletters" data-reactid="789">Newsletters</option></select></div><div class="SelectLinks__dropdown___24BA0" data-reactid="790"><select data-reactid="791"><option selected="" value="/about/us" data-reactid="792">About</option><option value="/about/careers" data-reactid="793">Careers</option><option value="/about/contact" data-reactid="794">Contact</option><option value="/about/faq" data-reactid="795">FAQ</option><option value="http://www.condenast.com/brands/the-new-yorker" data-reactid="796">Media Kit</option><option value="/about/press" data-reactid="797">Press</option><option value="/about/accessibility-help/" data-reactid="798">Accessibility Help</option></select></div></div><div class="SiteFooter__otherContent___1WRDN" data-reactid="799"><div class="FooterSocialLinks__socialLinks___30_Va" data-reactid="800"><h4 data-reactid="801">Follow Us</h4><div data-reactid="802"><a data-track-location="footer" data-track-type="socialFollow" data-source="facebook" href="https://www.facebook.com/newyorker/" target="_blank" data-reactid="803"><svg width="8" height="17" viewBox="0 0 8 17" xmlns="http://www.w3.org/2000/svg"><path d="M1.7 17h3.5V8.5h2.3l.3-3H5.2V4.2c0-.7 0-1 1.2-1h1.4V0H5.5C2.7 0 1.7 1.4 1.7 3.8v1.8H0v3h1.7V17z" fill="#ddd" fill-rule="evenodd"/></svg></a><a data-track-location="footer" data-track-type="socialFollow" data-source="twitter" href="https://twitter.com/NewYorker" target="_blank" data-reactid="804"><svg width="19" height="15" viewBox="0 0 19 15" xmlns="http://www.w3.org/2000/svg"><path d="M18.223 1.739c-.654.295-1.3565.4943-2.094.584.7527-.459 1.3308-1.1858 1.603-2.0518-.7045.425-1.4847.7336-2.3152.9C14.7518.4504 13.8043 0 12.7558 0c-2.0137 0-3.646 1.6605-3.646 3.7084 0 .2908.032.5737.0943.8452-3.03-.1547-5.7164-1.631-7.5146-3.8748-.3138.5478-.4936 1.1848-.4936 1.8644 0 1.2866.6436 2.4217 1.622 3.0868-.5977-.0192-1.16-.186-1.6515-.464l-.0003.0468c0 1.7968 1.2567 3.2957 2.9246 3.6364-.306.0847-.628.13-.9605.13-.235 0-.4632-.0232-.686-.0665.464 1.4733 1.8105 2.5455 3.406 2.5755-1.248.9946-2.82 1.5875-4.528 1.5875-.2943 0-.5845-.0176-.8698-.0518 1.6136 1.052 3.53 1.666 5.5888 1.666 6.706 0 10.3733-5.651 10.3733-10.5515 0-.1608-.0036-.3208-.0106-.4798.7123-.5228 1.3304-1.176 1.8192-1.9197" fill="#ddd" fill-rule="evenodd"/></svg></a><a data-track-location="footer" data-track-type="socialFollow" data-source="snapchat" href="https://www.snapchat.com/add/newyorkermag" target="_blank" data-reactid="805"><svg width="21" height="20" viewBox="0 0 21 20" xmlns="http://www.w3.org/2000/svg" xmlns:xlink="http://www.w3.org/1999/xlink"><defs><path id="a" d="M20.3 19H0V0h20.3z"/></defs><g fill="none" fill-rule="evenodd"><path d="M10.3 19H10c-1.2 0-2-.6-2.7-1.2-.5-.3-1-.7-1.5-.7H4c-.3 0-.4.2-.5.2-.2 0-.3 0-.3-.2v-.5l-.4-.6c-2-.4-2.4-.8-2.6-1v-.2s0-.2.2-.2c3-.5 4-3.4 4.2-3.6.2-.3.2-.6 0-1 0-.4-.7-.6-1.2-.7h-.2c-1-.4-1-.8-1-1 0-.2.5-.4.8-.4h.3l1 .2c.4 0 .6-.2.6-.2v-.6c-.2-1.4-.3-3 0-4 1.4-3 4-3 5-3h.3c.8 0 3.6 0 4.8 3 .6 1 .5 2.6.4 4V8s0 .2.5.2c.2 0 .5 0 1-.3 0 0 0-.2.2-.2h.3c.3.2.5.4.5.6 0 .2-.2.5-1 .8H17c-.4.2-1 .4-1.2 1v.8c0 .2 1.4 3 4.3 3.6v.2c0 .4-.6.8-2.5 1-.2 0-.2.3-.3.7v.5l-.4.2h-.4c-.3-.2-.7-.2-1-.2h-1c-.5 0-1 .5-1.5.8-.8.6-1.6 1-2.7 1" fill="#DDD"/><g transform="translate(0 .02)"><mask id="b" fill="#ddd"><use xlink:href="#a"/></mask><path d="M10.3 0H10C9 0 6.2.3 5 3.3c-.5 1-.4 2.7-.3 4V8h-.4c-.3 0-.6 0-1-.3H3c-.4 0-1 .2-1 .6 0 .3 0 .7 1 1l.4.2c.4 0 1 .3 1 .7.2.2.2.4 0 .7 0 0-1.3 3-4 3.3-.3 0-.4.3-.4.5v.2c.3.4 1 .8 2.7 1 0 0 0 .4.2.5v.5c0 .2 0 .4.5.4H4c.2-.2.5-.2 1-.2h.8c.5 0 1 .4 1.4.8.7.5 1.6 1 2.8 1h.3c1.2 0 2-.5 2.7-1 .5-.4 1-.7 1.5-.8h2.3c.3 0 .4 0 .5-.2v-.5l.2-.5c1.6-.2 2.5-.6 2.7-1v-.2c0-.2 0-.4-.3-.5C17 14 16 11 16 11c-.3-.4-.3-.6-.2-.8 0-.4.7-.6 1-.7h.4c.7-.4 1-.7 1-1 0-.4-.2-.7-.6-.8h-.4-.4l-1 .2h-.2v-.6c0-1.4.2-3-.3-4-1.3-3-4-3.4-5-3.4m0 .5c.8 0 3.4.2 4.7 3 .4 1 .3 2.5.2 3.8V8s.2.3.7.4c.2 0 .6 0 1-.3h.4l.4.4s0 .3-.8.6h-.2c-.5.2-1.2.5-1.4 1-.2.3 0 .7 0 1 .2.3 1.5 3.3 4.4 3.8 0 0-.3.5-2.3.8-.3 0-.4.4-.5.8v.5h-.2-.4c-.3-.2-.6-.2-1-.2h-1c-.6 0-1 .5-1.6.8-.7.5-1.4 1-2.5 1H10c-1 0-1.8-.5-2.5-1-.6-.3-1-.7-1.7-.8h-2-.3c-.2 0-.2-.2-.2-.4 0-.4-.2-.7-.5-.8-2-.3-2.3-.7-2.4-.8 3-.5 4.3-3.5 4.4-3.7.2-.3.2-.7 0-1 0-.5-.8-.8-1.3-1h-.3c-.6-.2-.8-.5-.7-.6 0-.2.3-.4.5-.4h.2c.4.3.8.4 1 .4.6 0 .8-.3 1-.3v-.7c-.2-1.3-.3-3 0-4C6.6.8 9.3.6 10 .6h.3" fill="#DDD" mask="url(#b)"/></g></g></svg></a><a data-track-location="footer" data-track-type="socialFollow" data-source="youtube" href="https://www.youtube.com/user/NewYorkerDotCom" target="_blank" data-reactid="806"><svg height=20px version=1.1 viewBox="0 0 29 20"width=29px xmlns=http://www.w3.org/2000/svg xmlns:xlink=http://www.w3.org/1999/xlink><defs><polygon id=path-1 points="28.4130459 9.99197775 28.4130459 19.9839499 0 19.9839499 0 9.99197775 0 5.56328223e-06 28.4130459 5.56328223e-06"></polygon></defs><g id=footer fill=none fill-rule=evenodd stroke=none stroke-width=1><g id=footer-1280 transform="translate(-852.000000, -607.000000)"><g id=FOLLOW transform="translate(850.000000, 528.000000)"><g id=YT transform="translate(2.000000, 79.000000)"><mask fill=white id=mask-2><use xlink:href=#path-1></use></mask><g id=Clip-2></g><path d="M11.2732128,13.6774465 L11.2719054,5.6941363 L18.949235,9.69969958 L11.2732128,13.6774465 Z M28.1293185,4.31113213 C28.1293185,4.31113213 27.8516551,2.35316273 26.9997775,1.49093741 C25.9193324,0.35925452 24.7082615,0.353663421 24.152879,0.287404729 C20.1768846,5.56328223e-06 14.2126843,5.56328223e-06 14.2126843,5.56328223e-06 L14.2003338,5.56328223e-06 C14.2003338,5.56328223e-06 8.2362726,5.56328223e-06 4.26013908,0.287404729 C3.70472879,0.353663421 2.49410292,0.35925452 1.41326843,1.49093741 C0.561363004,2.35316273 0.284144645,4.31113213 0.284144645,4.31113213 C0.284144645,4.31113213 0,6.6104089 0,8.90971349 L0,11.0652907 C0,13.3645396 0.284144645,15.6638164 0.284144645,15.6638164 C0.284144645,15.6638164 0.561363004,17.6218136 1.41326843,18.4840389 C2.49410292,19.615694 3.91385257,19.5799221 4.54620306,19.6985313 C6.81927677,19.9165007 14.2065369,19.9839555 14.2065369,19.9839555 C14.2065369,19.9839555 20.1768846,19.9749708 24.152879,19.6875438 C24.7082615,19.6213129 25.9193324,19.615694 26.9997775,18.4840389 C27.8516551,17.6218136 28.1293185,15.6638164 28.1293185,15.6638164 C28.1293185,15.6638164 28.4130459,13.3645396 28.4130459,11.0652907 L28.4130459,8.90971349 C28.4130459,6.6104089 28.1293185,4.31113213 28.1293185,4.31113213 L28.1293185,4.31113213 Z"fill=#DDDDDD id=Fill-1 mask=url(#mask-2)></path></g></g></g></g></svg></a><a data-track-location="footer" data-track-type="socialFollow" data-source="instagram" href="https://www.instagram.com/newyorkermag/" target="_blank" data-reactid="807"><svg height=21px version=1.1 viewBox="0 0 20 21"width=20px xmlns=http://www.w3.org/2000/svg xmlns:xlink=http://www.w3.org/1999/xlink><defs><polygon id=path-1 points="0 20.0102963 0 0.00631277204 20.0039835 0.00631277204 20.0039835 20.0102963"></polygon></defs><g id=footer fill=none fill-rule=evenodd stroke=none stroke-width=1><g id=footer-1280 transform="translate(-916.000000, -607.000000)"><g id=FOLLOW transform="translate(850.000000, 528.000000)"><g id=insta transform="translate(66.000000, 79.000000)"><g id=Group-3 transform="translate(0.000000, 0.033390)"><mask fill=white id=mask-2><use xlink:href=#path-1></use></mask><g id=Clip-2></g><path d="M10.0019917,0.00631277204 C7.28559403,0.00631277204 6.94498226,0.017826633 5.87820319,0.0665024727 C4.81356808,0.115059204 4.08652732,0.284154148 3.45028724,0.531384534 C2.79256786,0.786992247 2.23478086,1.12899362 1.67874078,1.6850337 C1.1227007,2.24111348 0.780699327,2.79890049 0.525091614,3.45661987 C0.277821524,4.09282024 0.108766283,4.81990071 0.0601698492,5.88449611 C0.0114940095,6.95131488 -1.98514845e-05,7.29192665 -1.98514845e-05,10.0083244 C-1.98514845e-05,12.7246824 0.0114940095,13.0652942 0.0601698492,14.1321129 C0.108766283,15.1967083 0.277821524,15.9237888 0.525091614,16.5599892 C0.780699327,17.2177086 1.1227007,17.7754956 1.67874078,18.3315753 C2.23478086,18.8876154 2.79256786,19.2296168 3.45028724,19.4852245 C4.08652732,19.7324549 4.81356808,19.9015498 5.87820319,19.9501066 C6.94498226,19.9987824 7.28559403,20.0102963 10.0019917,20.0102963 C12.7183895,20.0102963 13.0590012,19.9987824 14.1257803,19.9501066 C15.1904154,19.9015498 15.9174562,19.7324549 16.5536963,19.4852245 C17.2114156,19.2296168 17.7692026,18.8876154 18.3252427,18.3315753 C18.8812828,17.7754956 19.2232842,17.2177086 19.4788919,16.5599892 C19.726162,15.9237888 19.8952172,15.1967083 19.9438136,14.1321129 C19.9924895,13.0652942 20.0040033,12.7246824 20.0040033,10.0083244 C20.0040033,7.29192665 19.9924895,6.95131488 19.9438136,5.88449611 C19.8952172,4.81990071 19.726162,4.09282024 19.4788919,3.45661987 C19.2232842,2.79890049 18.8812828,2.24111348 18.3252427,1.6850337 C17.7692026,1.12899362 17.2114156,0.786992247 16.5536963,0.531384534 C15.9174562,0.284154148 15.1904154,0.115059204 14.1257803,0.0665024727 C13.0590012,0.017826633 12.7183895,0.00631277204 10.0019917,0.00631277204 M10.0019917,1.80847023 C12.6726516,1.80847023 12.9889652,1.81867389 14.0436349,1.86679389 C15.0188192,1.91126121 15.5484568,2.0742022 15.9009,2.21117744 C16.3677275,2.39262001 16.7009149,2.60935851 17.0509362,2.95937989 C17.4009179,3.30936156 17.6176961,3.64254887 17.7991387,4.10941608 C17.9360742,4.46185934 18.0990152,4.99145724 18.1435222,5.96664156 C18.1916422,7.02131122 18.2018062,7.33766447 18.2018062,10.0083244 C18.2018062,12.6789446 18.1916422,12.9952978 18.1435222,14.0499675 C18.0990152,15.0251518 17.9360742,15.5547497 17.7991387,15.907193 C17.6176961,16.3740602 17.4009179,16.7072475 17.0509362,17.0572292 C16.7009149,17.4072505 16.3677275,17.623989 15.9009,17.8054316 C15.5484568,17.9424068 15.0188192,18.1053478 14.0436349,18.1498152 C12.989124,18.1979351 12.6728105,18.2081388 10.0019917,18.2081388 C7.33117304,18.2081388 7.01485949,18.1979351 5.96034863,18.1498152 C4.98516432,18.1053478 4.45552671,17.9424068 4.10308346,17.8054316 C3.63625595,17.623989 3.30306864,17.4072505 2.95304726,17.0572292 C2.60306559,16.7072475 2.38628738,16.3740602 2.20484482,15.907193 C2.06790928,15.5547497 1.90496829,15.0251518 1.86046127,14.0499675 C1.81234127,12.9952978 1.80217731,12.6789446 1.80217731,10.0083244 C1.80217731,7.33766447 1.81234127,7.02131122 1.86046127,5.96664156 C1.90496829,4.99145724 2.06790928,4.46185934 2.20484482,4.10941608 C2.38628738,3.64254887 2.60306559,3.30936156 2.95304726,2.95937989 C3.30306864,2.60935851 3.63625595,2.39262001 4.10308346,2.21117744 C4.45552671,2.0742022 4.98516432,1.91126121 5.96030893,1.86679389 C7.0150183,1.81867389 7.33133185,1.80847023 10.0019917,1.80847023"fill=#DDDDDD id=Fill-1 mask=url(#mask-2)></path></g><path d="M10.0019917,13.375688 C8.16068716,13.375688 6.66797465,11.8830152 6.66797465,10.0417106 C6.66797465,8.20036631 8.16068716,6.7076935 10.0019917,6.7076935 C11.8432963,6.7076935 13.3359691,8.20036631 13.3359691,10.0417106 C13.3359691,11.8830152 11.8432963,13.375688 10.0019917,13.375688 M10.0019917,4.90553604 C7.16537344,4.90553604 4.86581719,7.20505259 4.86581719,10.0417106 C4.86581719,12.8783289 7.16537344,15.1778455 10.0019917,15.1778455 C12.8386101,15.1778455 15.1381663,12.8783289 15.1381663,10.0417106 C15.1381663,7.20505259 12.8386101,4.90553604 10.0019917,4.90553604"fill=#DDDDDD id=Fill-4></path><path d="M16.5413208,4.70260623 C16.5413208,5.36548699 16.0039412,5.90282697 15.3411001,5.90282697 C14.6782193,5.90282697 14.1408396,5.36548699 14.1408396,4.70260623 C14.1408396,4.03972546 14.6782193,3.50238548 15.3411001,3.50238548 C16.0039412,3.50238548 16.5413208,4.03972546 16.5413208,4.70260623"fill=#DDDDDD id=Fill-6></path></g></g></g></g></svg></a></div></div><a class="SiteFooter__condeLogo___2ThbJ" href="http://www.condenast.com/" target="_blank" data-reactid="808"><svg xmlns="http://www.w3.org/2000/svg" width="120" height="25" viewBox="0 120 288 50"><style>.a { fill:#999; }</style><path d="M26 161c-1.5 1.4-3.6 2.4-5.8 2.4-7 0-8.5-6-8.5-14s1.5-14 8.4-14c2.3 0 4.3 1 5.8 2.4l3.8 6.7h.2v-9.2h-.2l-3.6 2c-1-1-3-2.4-6-2.4-6.6 0-14.5 5.2-14.5 14.4S13.3 164 20.2 164c2.7 0 5-1.5 6-2.4l3.8 2h.2v-9.7H30l-4 7zm22.5-26.2c-6.5 0-14.2 5.7-14.2 14.5s7.7 14.5 14.2 14.5 14.2-5.7 14.2-14.5-7.7-14.5-14.2-14.5zm0 28.6c-6.6 0-8-5.6-8-14s1.4-14 8-14 8 5.6 8 14-1.3 14-8 14zm37-27.6h3.6v18l-12.4-18.4H66v.4h3.4V163H66v.4h7.5v-.4H70v-27.2l19.5 28h.2v-28h3.5v-.4h-7.6m22.8 0h-12v.4h3.4V163h-3.5v.4h12c6 0 16.3-2.7 16.3-14.2s-10.2-13.8-16.2-13.8zm-.2 27.6h-3v-27.2h3c7 0 10.4 5 10.4 13.6 0 8.5-3.3 13.6-10.4 13.6zm38.8 0h-10.3v-14.4h5.2c1.6 1.4 3.6 5.4 3.6 5.4h.2v-11.3h-.2s-2 4-3.7 5.4h-5.3V136h9.2c2.6 1.8 5 7 5 7v-7.4h-23.3v.4h3.5V163h-3.5v.4H152V154h-.3s-2.4 6.3-4.7 9zm-.5-33.6l-.7-2 .6-3.2h-.2l-8.6 7.2v.2m64.4 3.8h-7.6v.4h3.5v18l-12.6-18.4h-10.7v.4h3.5V163h-3.5v.4h7.6v-.4h-3.5v-27.2l19.5 28h.2v-28h3.5m14.7-1.4h-.2L204.8 163h-3.2v.4h7.3v-.4h-3.6l4.4-10.7h8.8l4.6 10.7h-3.5v.4h12.8v-.4h-3.3l-12.5-28.6zm-6.8 17.4l4-10 4.3 10H210zm38.2-5L243 145c-2.8-1-5-2.3-5-5 0-3 2.4-4.7 6-4.7 2.2 0 3.6.6 5 2l3.7 6.4h.2v-9h-.3l-3.6 2c-.8-1-2.4-2-5-2-5.2 0-8.6 3.6-8.6 8.5 0 4 2.7 6 5.8 7.2l5.5 2c3 1 4.8 2.4 4.8 5.3 0 2.8-1.8 5.6-6.5 5.6-2.2 0-4.8-1-6.2-2.5l-4-7.2h-.3l.2 9.7h.2l3.6-2c1 1 3.7 2.4 6.4 2.4 6 0 9.8-4.3 9.8-9 0-4.5-3.2-6.6-6.6-7.8zm9.2-11.4v8.7h.2s2.6-5.5 5-8.2h4.6V163H263v.4h14v-.4h-4.4v-27.2h4.7c2.4 2.7 5 8.3 5 8.3v-8.6h-25z" class="a"/></svg></a><div class="FooterLegalDisclaimer__legalDisclaimer___DJ3BV" data-reactid="809"><span role="img" aria-label="copyright icon" data-reactid="810">©</span><!-- react-text: 811 --> <!-- /react-text --><!-- react-text: 812 -->2018<!-- /react-text --><!-- react-text: 813 --> Condé Nast. All rights reserved. Use of this site constitutes acceptance of our <!-- /react-text --><a class="Link__link___3dWao  " href="http://www.condenast.com/privacy-policy" data-reactid="814"><!-- react-text: 815 -->user agreement<!-- /react-text --></a><!-- react-text: 816 --> (effective 1/2/2016) and <!-- /react-text --><a class="Link__link___3dWao  " href="http://www.condenast.com/privacy-policy#privacypolicy" data-reactid="817"><!-- react-text: 818 -->privacy policy<!-- /react-text --></a><!-- react-text: 819 --> (effective 1/2/2016). <!-- /react-text --><a class="Link__link___3dWao  " href="http://www.condenast.com/privacy-policy#privacypolicy-california" data-reactid="820"><!-- react-text: 821 -->Your California privacy rights<!-- /react-text --></a><!-- react-text: 822 -->. The material on this site may not be reproduced, distributed, transmitted, cached or otherwise used, except with prior written permission of Condé Nast. <!-- /react-text --><i data-reactid="823">The New Yorker</i><!-- react-text: 824 --> may earn a portion of sales from products and services that are purchased through links on our site as part of our affiliate partnerships with retailers.<!-- /react-text --></div><div class="SiteFooter__sisterSitesContainer___1A6Co" data-reactid="825"><div class="SelectLinks__dropdown___24BA0" data-reactid="826"><select data-reactid="827"><option selected="" value="" disabled="" data-reactid="828">Our sites</option><option value="http://www.allure.com" data-reactid="829">Allure</option><option value="http://arstechnica.com" data-reactid="830">Ars Technica</option><option value="http://www.backchannel.com" data-reactid="831">Backchannel</option><option value="http://www.bonappetit.com" data-reactid="832">Bon Appétit</option><option value="http://www.brides.com" data-reactid="833">Brides</option><option value="http://www.cntraveler.com" data-reactid="834">Condé Nast Traveler</option><option value="http://www.epicurious.com" data-reactid="835">Epicurious</option><option value="http://www.glamour.com" data-reactid="836">Glamour</option><option value="http://www.golfdigest.com" data-reactid="837">Golf Digest</option><option value="http://www.golfdigest.com/golfworld/" data-reactid="838">Golf World</option><option value="http://www.gq.com" data-reactid="839">GQ</option><option value="http://www.gq.com/gqstyle" data-reactid="840">GQ Style</option><option value="http://pitchfork.com" data-reactid="841">Pitchfork</option><option value="http://www.self.com" data-reactid="842">Self</option><option value="http://www.style.com" data-reactid="843">Style</option><option value="http://www.teenvogue.com" data-reactid="844">Teen Vogue</option><option value="http://www.newyorker.com" data-reactid="845">The New Yorker</option><option value="http://thescene.com" data-reactid="846">The Scene</option><option value="http://www.vanityfair.com" data-reactid="847">Vanity Fair</option><option value="http://www.vogue.com" data-reactid="848">Vogue</option><option value="http://www.wmagazine.com" data-reactid="849">W</option><option value="http://www.wired.com" data-reactid="850">Wired</option></select></div></div></div></div></div></footer></main><div class="Drawer__drawer___2bwvY" data-reactid="851"><div class="Drawer__menu___1vYCL" data-reactid="852"><div class="SearchInput__modalSearchInput___31Bpr " data-reactid="853"><form data-reactid="854"><div data-reactid="855"><input type="text" id="search-input-menu" data-reactid="856"/><div class="SearchInput__modalIcon___YSkrm" data-reactid="857"><i class="SearchInput__modalIcon___YSkrm" data-reactid="858"><svg viewBox="0 0 18 18" xmlns="http://www.w3.org/2000/svg"><path d="M2 7.5C2 4.468 4.468 2 7.5 2S13 4.468 13 7.5 10.532 13 7.5 13 2 10.532 2 7.5zm16 9.086l-4.545-4.545C14.42 10.779 15 9.208 15 7.5 15 3.364 11.636 0 7.5 0S0 3.364 0 7.5 3.364 15 7.5 15c1.708 0 3.279-.58 4.541-1.545L16.586 18 18 16.586z" fill="#121212" fill-rule="evenodd" /></svg></i></div></div></form></div><div class="Navigation__drawerMenu___2KwE1" data-reactid="859"><ul data-reactid="860"><li data-track-type="nav" data-track-location="button-news" data-reactid="861"><a class="Link__link___3dWao  " href="/news" data-reactid="862"><!-- react-text: 863 -->News<!-- /react-text --></a></li><li data-track-type="nav" data-track-location="button-culture" data-reactid="864"><a class="Link__link___3dWao  " href="/culture" data-reactid="865"><!-- react-text: 866 -->Culture<!-- /react-text --></a></li><li data-track-type="nav" data-track-location="button-books" data-reactid="867"><a class="Link__link___3dWao  " href="/books" data-reactid="868"><!-- react-text: 869 -->Books<!-- /react-text --></a></li><li data-track-type="nav" data-track-location="button-business-&amp;-tech" data-reactid="870"><a class="Link__link___3dWao  " href="/business" data-reactid="871"><!-- react-text: 872 -->Business &amp; Tech<!-- /react-text --></a></li><li data-track-type="nav" data-track-location="button-humor" data-reactid="873"><a class="Link__link___3dWao  " href="/humor" data-reactid="874"><!-- react-text: 875 -->Humor<!-- /react-text --></a></li><li data-track-type="nav" data-track-location="button-cartoons" data-reactid="876"><a class="Link__link___3dWao  " href="/cartoons" data-reactid="877"><!-- react-text: 878 -->Cartoons<!-- /react-text --></a></li><li data-track-type="nav" data-track-location="button-magazine" data-reactid="879"><a class="Link__link___3dWao  " href="/magazine" data-reactid="880"><!-- react-text: 881 -->Magazine<!-- /react-text --></a></li><li data-track-type="nav" data-track-location="button-video" data-reactid="882"><a class="Link__link___3dWao  " href="http://video.newyorker.com" data-reactid="883"><!-- react-text: 884 -->Video<!-- /react-text --></a></li><li data-track-type="nav" data-track-location="button-podcasts" data-reactid="885"><a class="Link__link___3dWao  " href="/podcast" data-reactid="886"><!-- react-text: 887 -->Podcasts<!-- /react-text --></a></li><li data-track-type="nav" data-track-location="button-archive" data-reactid="888"><a class="Link__link___3dWao  " href="/archive" data-reactid="889"><!-- react-text: 890 -->Archive<!-- /react-text --></a></li><li data-track-type="nav" data-track-location="button-goings-on" data-reactid="891"><a class="Link__link___3dWao  " href="/goings-on-about-town/" data-reactid="892"><!-- react-text: 893 -->Goings On<!-- /react-text --></a></li><li class="Navigation__subscribe___3OCs_" data-reactid="894"><!-- react-empty: 895 --></li></ul><!-- react-empty: 896 --></div><div class="UserInfo__drawerMenu___2v2Tk" data-reactid="897"><div class="UserInfo__userInfoItems___2TxZh" data-reactid="898"><a class="Link__link___3dWao UserInfo__shop___20W-I " href="http://www.condenaststore.com/~/new-yorker-store?utm_medium=referral&amp;utm_source=NewYorker&amp;utm_content=SHOP&amp;AID=1247905545" data-reactid="899"><!-- react-text: 900 -->Shop<!-- /react-text --></a><div class="SignIn__signIn___1aO_Z" data-reactid="901"><a class="Link__link___3dWao  " href="https://account.newyorker.com" data-reactid="902"><!-- react-text: 903 -->Sign in<!-- /react-text --></a><a class="Link__link___3dWao  " href="https://account.newyorker.com/linksubscription" data-reactid="904"><!-- react-text: 905 -->Link your subscription<!-- /react-text --></a></div></div><div data-track-type="nav" data-track-location="search-icon" class="UserInfo__search___1oSUo" data-reactid="906"><div class="Lightbox__lightbox___2lLZl Lightbox__white___jj_9p  " role="button" tabindex="0" data-reactid="907"><div class="" data-reactid="908"><a class="Link__link___3dWao SearchModal__icon___2cTZE " data-reactid="909"><div class="Link__content___6CYL1" data-reactid="910"><i class="Link__icon___3s7us Link__left___1QLUQ " data-reactid="911"><svg viewBox="0 0 18 18" xmlns="http://www.w3.org/2000/svg"><path d="M2 7.5C2 4.468 4.468 2 7.5 2S13 4.468 13 7.5 10.532 13 7.5 13 2 10.532 2 7.5zm16 9.086l-4.545-4.545C14.42 10.779 15 9.208 15 7.5 15 3.364 11.636 0 7.5 0S0 3.364 0 7.5 3.364 15 7.5 15c1.708 0 3.279-.58 4.541-1.545L16.586 18 18 16.586z" fill="#121212" fill-rule="evenodd" /></svg></i></div></a></div></div></div></div></div><div class="Drawer__overlay___TrAm9" role="button" tabindex="0" data-reactid="912"></div></div></div></div></div><script>window.__TNY__.INITIAL_STATE = {"primary":{"ads":{"channel":"Culture","subChannel":"Annals of Gastronomy"},"analytics":{"appVersion":"1.0","brand":"The New Yorker","canonical":"https://www.newyorker.com/culture/annals-of-gastronomy/the-first-time-i-met-paul-bocuse","contentID":"5a6602e016b0267062e903f3","contentLength":3,"contentSource":"web","contributor":"Bill Buford","display":"The First Time I Met Paul Bocuse","env":"production","keywords":{"display":"Food|Postscript|Cooking|France"},"magazineTOCSection":"","modifiedDate":"2018-01-23T00:37:42.779Z","numOfSearchResults":0,"pageType":"Post","pageValue":"page_1","publishDate":"2018-01-22T19:08:28-05:00","searchTerms":"","section":"Culture","subsection":"Annals of Gastronomy","syndication":"","syndicationUrl":""},"body":["div",["inline-embed",{"type":"section","meta":"%7B%22type%22%3A%22section%22%2C%22body%22%3A%22%22%7D","ref":""}],["inline-embed",{"type":"section","meta":"%7B%22type%22%3A%22section%22%2C%22body%22%3A%22%3Cp%3EUntil%20%3Cinline-embed%20type%3D%5C%22analytics%5C%22%20meta%3D%5C%22%257B%2522type%2522%253A%2522analytics%2522%252C%2522url%2522%253A%25221%2522%252C%2522width%2522%253A%2522%2522%252C%2522height%2522%253A%2522%2522%252C%2522caption%2522%253A%2522%2522%257D%5C%22%20ref%3D%5C%22%5C%22%3E%3C%2Finline-embed%3E%5Cn1985%20or%20thereabouts%2C%20Paul%20Bocuse%20was%20a%20very%20famous%20chef.%20He%20was%20a%5Cnhousehold%20name%2C%20with%20an%20image%20so%20widely%20disseminated%20that%20just%20about%5Cneveryone%20could%20tell%20you%20what%20he%20looked%20like%E2%80%94the%20big%20lips%2C%20the%20toque%2C%20and%5Cnholding%20a%20live%20%3Cem%3Epoulet%20de%20Bresse%3C%2Fem%3E%20in%20his%20arms%2C%20say%2C%20and%20petting%20its%5Cnhead%E2%80%94even%20if%20few%20people%20had%20any%20idea%20what%20his%20food%20might%20be%20like.%20Then%2C%5Cnaround%201985%20or%20so%2C%20Bocuse%20turned%20into%20an%20icon.%20One%20moment%3A%20famous%20guy%20at%5Cnstove.%20Then%3A%20Pope%20of%20restaurant%20people.%20He%20became%20the%20undisputed%5Cnemissary%20of%20the%20kitchen%20mission.%20He%20became%20Frenchness.%3C%2Fp%3E%5Cn%3Cp%3EIt%20is%20not%20entirely%20clear%20how%20this%20happened%2C%20because%20(and%20Bocuse%20would%5Cnhave%20agreed)%20there%20were%2C%20and%20are%2C%20more%20talented%20and%20savvy%20chefs%20around.%5CnHe%20never%20hosted%20a%20food%20show.%20Although%20a%20master%20of%20the%20photo%20op%20(hamming%5Cnit%20up%20in%20drag%2C%20stripping%20down%20to%20show%20off%20his%20tattoo%20of%20a%20Bresse%5Cnchicken%2C%20straddling%20his%20Harley%20Davidson)%2C%20he%20rarely%20appeared%20on%5Cntelevision.%20He%20published%20cookbooks.%20None%20was%20a%20game-changer.%20Apart%20from%5Cntwo%20forays%20abroad%E2%80%94one%20in%20Japan%20(which%20failed)%20and%20another%20at%20Walt%20Disney%5CnWorld%2C%20in%20Florida%20(which%20thrives%2C%20run%20by%20his%20son%20J%C3%A9r%C3%B4me)%E2%80%94he%20never%20really%5Cnfranchised%20his%20name.%20And%20yet%2C%20in%20the%20mysterious%2C%20ineffable%20way%20of%20these%5Cnthings%2C%20Bocuse%2C%20who%20died%20on%20Saturday%2C%20at%20the%20age%20of%20ninety-one%2C%20had%5Cnsomething%20that%20no%20other%20chef%20has%20in%20the%20same%20abundance%3A%20an%20undeniably%5Cninfectious%20culinary%20charisma.%20Bocuse%20is%20what%20people%20want%20a%20grand%20French%5Cnchef%20to%20be.%3C%2Fp%3E%5Cn%3Cp%3EIn%202009%2C%20I%20moved%20to%20Lyon%20with%20my%20family%2C%20to%20train%20to%20be%20a%20French%20chef%5Cn(not%20the%20%E2%80%9Cgrand%E2%80%9D%20kind%2C%20but%20more%20like%20a%20garden%20variety%20kitchen%20Frenchie).%5CnWe%20stayed%20five%20years.%20Bocuse%20was%20everywhere%20and%20nowhere.%20I%20enrolled%20in%20a%5Cncooking%20school%E2%80%94%3Cem%3Ethe%3C%2Fem%3E%20cooking%20school%2C%20l%E2%80%99Institut%20Paul%20Bocuse%E2%80%94where%20just%5Cnabout%20every%20day%20there%20was%20a%20rumor%20that%20%E2%80%9CMonsieur%20Paul%E2%80%9D%20(as%20he%20is%5Cnreferred%20to%20locally)%20was%20going%20to%20make%20an%20appearance.%20He%20never%20did%2C%5Cnalthough%20he%20regularly%20interrupted%20the%20instructors%2C%20calling%20them%20on%20their%5Cncell%20phones%20during%20class.%20I%20was%20given%20hundreds%20of%20recipes%20and%20expected%5Cnto%20memorize%20them%3B%20they%20were%20mainly%20%3Cem%3Ehis%3C%2Fem%3E%20recipes%2C%20I%20would%20eventually%5Cnlearn.%20Around%20the%20city%20there%20were%20three%20murals%20devoted%20to%20his%20image%2C%5Cnincluding%20one%20that%20was%20directly%20across%20the%20street%20from%20our%20apartment%20on%5Cnthe%20Saone%20river%3A%20he%20was%20what%20I%20looked%20at%20drinking%20my%20morning%20coffee.%5CnBridges%20were%20named%20after%20him.%20Les%20Halles%2C%20the%20indoor%20market%2C%20was%5Cnofficially%20Les%20Halles%20Paul%20Bocuse.%20And%20so%2C%20too%2C%20was%20the%20world%E2%80%99s%20biggest%5Cnand%20most%20hysterical%20cooking%20competition%2C%20the%20biennial%20Bocuse%20d%E2%80%99Or.%3C%2Fp%3E%5Cn%3Cp%3EThe%20event%20is%20organized%20like%20a%20World%20Cup%20(forty-eight%20nations%20competing%2C%5Cneach%20represented%20by%20a%20team%20of%20two%2C%20a%20chef%20and%20a%20deputy)%20and%20run%20like%20a%5Cndog%20show%2C%20with%20dishes%20paraded%20through%20a%20stadium%20in%20front%20of%20twenty-six%5Cnjudges.%20It%20culminates%20in%20an%20awards%20ceremony%20that%20manages%20to%20use%20the%5Cniconography%20of%20the%20Olympics%20(with%20a%20podium%20for%20gold%2C%20silver%2C%20and%5Cnbronze)%2C%20the%20Oscars%20(the%20%E2%80%9CBocuse%20d%E2%80%99Or%E2%80%9D%20is%20a%20gold%20statuette%20in%20the%20image%5Cnof%20Paul%20Bocuse)%2C%20and%20a%20New%20York%20City%20bar%20mitzvah%20(strobe%20lights%2C%20loud%5Cnoom-pah-pah%20music%2C%20and%20a%20ceiling%20drop%20of%20gold%20confetti).%20It%20is%20%E2%80%9Ctacky%E2%80%9D%5Cnmeets%20%E2%80%9Ctechnique%2C%E2%80%9D%20but%2C%20despite%20the%20tackiness%2C%20the%20%E2%80%9Ctechnique%E2%80%9D%20is%20real%3A%5Cnon%20display%20is%20the%20flashiest%2C%20most%20accomplished%20food%20on%20the%20planet.%3C%2Fp%3E%5Cn%3Cp%3EI%20attended%20my%20first%20Bocuse%20d%E2%80%99Or%20during%20our%20first%20year%20in%20France.%20I%5Cnshowed%20up%20at%209%20%3Cem%20class%3D%5C%22small%5C%22%3EA.M.%3C%2Fem%3E%20on%20the%20first%20day%20of%20what%20would%20be%20a%20two-day%20competition.%20The%20chefs%20had%20been%20there%20since%20five.%20Each%20two-member%5Cnteam%2C%20plus%20a%20local%20%3Cem%3Ecommis%3C%2Fem%3E%2C%20a%20slave-helper%20provided%20by%20l%E2%80%99Institut%20Paul%5CnBocuse%2C%20was%20crushed%20bumpingly%20into%20a%20mini-kitchen%20cubicle%20the%20size%20of%20a%5Cnchanging%20room%20at%20a%20bad%20beach%20resort.%20The%20cubicles%2C%20side%20by%20side%2C%20were%5Cnspread%20out%20in%20a%20row%2C%20from%20left%20to%20right.%20They%20represented%20the%20%E2%80%9Cevent.%E2%80%9D%5CnIt%20was%20more%20rodeo%20arena%20than%20theatre.%20Facing%20the%20action%20were%20bleachers.%5CnThe%20mood%20in%20the%20mini-kitchen%20cubicles%20was%20of%20adrenaline%20and%20stress%20and%5Cnsweat.%20Every%20cook%20was%20intensely%20aware%20of%20a%20clock%20ticking.%20They%20were%5Cnfocussed%20and%20very%20quiet.%20The%20bleachers%20were%20very%20not.%20There%20were%5Cnthousands%20of%20seats%2C%20and%20they%20were%20already%20filled.%20Many%20spectators%5Cnwere%20in%20regional%20costume%20(the%20Japanese%20as%20samurais%2C%20the%20Mexicans%20in%5Cnsombreros)%20or%20draped%20in%20national%20flags%20(the%20Swedes%2C%20the%20Americans).%5CnThere%20was%20a%20very%20loud%20mariachi%20band%2C%20a%20very%20loud%20drum-and-bugle%20corps%2C%5Cna%20very%20loud%20pit%20percussion%20team%2C%20some%20guys%20banging%20cymbals%2C%20and%20many%5Cnmorons%20with%20stadium%20air%20horns.%20There%20wasn%E2%80%99t%20anything%20obvious%20to%20be%5Cnthundering%20about%3A%20no%20eviscerating%20of%20live%20goats%2C%20no%20fist%20pumping%2C%20no%5Cnchef%20standing%20up%20and%20saying%2C%20%E2%80%9CBam!%E2%80%9D%20But%20the%20cheering%20and%20the%20chanting%5Cnnever%20stopped.%3C%2Fp%3E%5Cn%3Cp%3EI%20felt%20uninformed%20and%20na%C3%AFve%E2%80%94how%20could%20I%20have%20not%20known%20that%20this%20kind%20of%5Cncooking%20(hunched-over%2C%20tortured%20people%20manipulating%20little%20things%20in%5Cnlittle%20ways)%20was%20a%20nationalist%20sport%3F%E2%80%94when%2C%20lo%20and%20behold%2C%20Bocuse%5Cnappeared.%3C%2Fp%3E%5Cn%3Cp%3EHe%20had%20turned%20up%20backstage%20and%20taken%20it%20upon%20himself%E2%80%94the%20toque%2C%20the%5Cnwhites%2C%20in%20a%20collar%20that%20looked%20like%20a%20French%20flag%E2%80%94to%20walk%20across%20the%5Cnfloor.%20He%20padded%20by%20me%2C%20softly%20and%20stiffly%2C%20imposingly%20tall%20in%20his%20chef%5Cnposture%2C%20doing%20small%20papal%20waves%2C%20and%20seemed%20not%20to%20notice%20that%20a%20queue%5Cnof%20competitors%20was%20forming%20behind%20him.%20They%20were%20Japanese%2C%20Korean%2C%5CnIndian%2C%20Brazilian%2C%20Norwegian%20.%20.%20.%20actually%2C%20I%20have%20no%20idea%20where%20they%5Cnwere%20from%2C%20except%20that%20there%20were%20lots%20of%20them%2C%20and%20they%20seemed%20not%20to%5Cnbe%20speaking%20either%20French%20or%20English.%20Only%20a%20moment%20before%2C%20they%20were%5Cnlocked%20down%3Cinline-embed%20type%3D%5C%22analytics%5C%22%20meta%3D%5C%22%257B%2522type%2522%253A%2522analytics%2522%252C%2522url%2522%253A%25222%2522%252C%2522width%2522%253A%2522%2522%252C%2522height%2522%253A%2522%2522%252C%2522caption%2522%253A%2522%2522%257D%5C%22%20ref%3D%5C%22%5C%22%3E%3C%2Finline-embed%3E%20in%20their%20panicky%20prep%2C%20when%20one%20of%20them%20noticed%20that%20Bocuse%5Cnwas%20in%20the%20house%20and%20abandoned%20his%20station.%20He%20seemed%20to%20then%20realize%2C%20Hey%2C%5Cnit%E2%80%99s%20O.K.%20to%20follow%20the%20great%20man%20around%E2%80%94or%20at%20least%20he%20registered%20that%5Cnno%20one%20was%20making%20him%20stop%E2%80%94and%20called%20to%20the%20others%20in%20their%5Cnmini-cubicle%20kitchens%2C%20and%20urged%20them%20to%20join%20in.%20The%20line%20quickly%5Cnbecame%20much%20longer%2C%20conga-style.%3C%2Fp%3E%5Cn%3Cp%3ENo%20one%20really%20knew%20what%20to%20do%20next%E2%80%94you%20can%20only%20get%20away%20with%20doing%20the%5Cnconga%20behind%20Bocuse%20for%20so%20long%E2%80%94when%20someone%20touched%20the%20great%20man%5Cnand%20dropped%20out%20of%20the%20queue%2C%20satisfied%20by%20the%20contact.%20The%20next%20person%5Cnin%20the%20queue%20touched%20a%20sleeve.%20Then%20it%20was%20a%20shoulder%2C%20the%20back%20of%5CnBocuse%E2%80%99s%20hand.%20An%20Asian%20chef%20was%20next%20up%2C%20and%20he%20seized%20Bocuse%E2%80%99s%20apron%5Cnby%20the%20hem%2C%20released%20it%2C%20and%20held%20the%20hand%20that%20had%20done%20the%20seizing%20by%5Cnits%20wrist%2C%20staring%20at%20it%2C%20shaking%20it%2C%20and%20screamed%20with%20wide-eyed%5Cninsanity%20as%20though%20his%20skin%20were%20burning.%20A%20cook%20fell%20to%20his%20knees%2C%20and%5Cnkissed%20the%20ground%20where%20the%20chef%20had%20tread%20(which%2C%20I%20don%E2%80%99t%20know%2C%20call%20me%5Cnprudish%2C%20seemed%20a%20little%20excessive).%3C%2Fp%3E%5Cn%3Cp%3EIt%20got%20grabby.%20The%20line%20lost%20its%20linear%20definition%2C%20and%2C%20just%20as%20it%20was%5Cnseeming%20outright%20dangerous%2C%20Bocuse%20was%20gone.%20Fearing%20a%20mob%20moment%2C%5Cnhandlers%20had%20emerged%20and%20ushered%20him%20through%20a%20backstage%20door.%3C%2Fp%3E%5Cn%3Cp%3EFor%20the%20rest%20of%20the%20year%2C%20it%20became%20my%20mission%20to%20meet%20the%20great%20man%5Cnproperly.%20I%20was%20told%20that%20I%20could%20reach%20him%20through%20his%20people.%20A%20wife%2C%5Cna%20daughter%2C%20a%20son.%20I%20wrote%20them.%20I%20phoned.%20It%20was%20the%20daughter%20who%5Cnfinally%20got%20back%20to%20me%2C%20and%20there%20was%20disdain%20in%20her%20reply%3A%20How%20dare%20you%5Cnthink%20that%20you%2C%20a%20lowly%20writer%20guy%2C%20would%20have%20anything%20to%20say%20to%5CnMonsieur%20Bocuse%20that%20is%20even%20remotely%20worth%20his%20time%3F%3C%2Fp%3E%5Cn%3Cp%3EThen%20I%20got%20lucky.%20I%20happened%20to%20be%20describing%20my%20frustrations%20in%20the%20bar%5Cnof%20La%20M%C3%A8re%20Brazier%2C%20which%20is%20among%20the%20restaurants%20where%20Bocuse%5Cntrained.%20Mathieu%20Viannay%20is%20the%20chef%2C%20and%20he%20was%20standing%20nearby.%20What%20I%5Cndidn%E2%80%99t%20know%20is%20that%20Viannay%20was%20also%20a%20member%20of%20an%20%E2%80%9Cinner%20circle.%E2%80%9D%20Its%5Cnmembers%E2%80%94basically%2C%20kick-ass%20chefs%20who%20understand%20the%20soul%20and%20suffering%5Cnof%20Lyonnais%20cooking%E2%80%94were%20awarded%20permanent%20access%20to%20the%20great%20man.%3C%2Fp%3E%5Cn%3Cp%3EViannay%20interrupted%20my%20conversation.%20%E2%80%9CYou%20want%20to%20meet%20Monsieur%20Paul%3F%E2%80%9D%3C%2Fp%3E%5Cn%3Cp%3E%E2%80%9CYes%2C%E2%80%9D%20I%20said%2C%20%E2%80%9Cvery%20much.%E2%80%9D%3C%2Fp%3E%5Cn%3Cp%3E%E2%80%9CCome%20by%20tomorrow%20morning%20at%20seven.%20Do%20not%20be%20late.%E2%80%9D%3C%2Fp%3E%5Cn%3Cp%3EViannay%20was%20waiting%20for%20me%20in%20his%20car.%20We%20crossed%20the%20Rh%C3%B4ne%20River%20and%5Cninto%20the%20Third%20Arrondissement%20and%20in%20the%20direction%20of%20Les%20Halles%20de%20Lyon%5CnPaul%20Bocuse.%20I%20then%20spotted%20the%20great%20chef%E2%80%99s%20vehicle%2C%20a%20massive%20black%5CnJeep%20Wrangler%20(American)%2C%20with%20all%20the%20doo-dahs%2C%20spotlights%20on%20the%20roof%2C%5Cna%20winch%20on%20the%20front%20bumper%2C%20and%20wide%20off-road%20tires%20(Michelin%2C%20of%5Cncourse).%20It%20was%20parked%20slapdash%20across%20a%20sidewalk%2C%20in%20front%20of%20the%5Cnmarket-hall%20entrance.%20The%20sight%20confirmed%20two%20Lyonnais%20rumors%3A%20one%2C%20that%5CnBocuse%20always%20had%20his%20early-morning%20coffee%20at%20Les%20Halles%3B%20and%2C%20two%2C%20that%5Cnthe%20police%20knew%20his%20vehicle%20and%20never%20ticketed%20it.%3C%2Fp%3E%5Cn%3Cp%3EFor%20our%20part%2C%20we%20used%20a%20conventional%20parking%20lot%2C%20paid%20for%20it%2C%20entered%5Cnthe%20hall%2C%20and%20stopped%20by%20a%20shellfish%20vendor%20named%20Leon%2C%20who%20gave%20us%20a%5Cnplatter%20of%20oysters.%20(%E2%80%9CLeon%20is%20the%20one%20you%20buy%20your%20oysters%20from%2C%E2%80%9D%5CnViannay%20told%20me.%20%E2%80%9C%3Cem%3ED%E2%80%99accord%3F%3C%2Fem%3E%E2%80%9D)%3C%2Fp%3E%5Cn%3Cp%3EWe%20made%20our%20way%20to%20Le%20Boulanger%2C%20an%20informal%20caf%C3%A9%20opposite%20the%20stand%20of%5CnLa%20M%C3%A8re%20Richard%2C%20the%20city%E2%80%99s%20famous%20cheese%20lady%2C%20where%20Viannay%20paused%20to%5Cnpoint%20out%20the%20quality%20of%20what%20was%20on%20display.%20(%E2%80%9CThis%20is%20where%20you%20will%5Cnbuy%20your%20cheese.%20%3Cem%3ED%E2%80%99accord%3C%2Fem%3E%3F%E2%80%9D)%3C%2Fp%3E%5Cn%3Cp%3EI%20looked%20across%20the%20aisle.%20There%20he%20was%3A%20Paul%20Bocuse%2C%20sitting%20by%5Cnhimself%2C%20finishing%20a%20coffee.%3C%2Fp%3E%5Cn%3Cp%3EHe%20was%20slumped%20slightly%20in%20his%20chair%2C%20and%2C%20wearing%20a%20black%20Pringle%5Cncotton%20shirt%2C%20a%20worker%E2%80%99s%20cotton%20jacket%2C%20black%20trousers%2C%20and%20sneakers%2C%20he%5Cnlooked%20like%20a%20regional%20train%20conductor%20at%20the%20end%20of%20his%20shift.%20On%5Cnspotting%20us%2C%20he%20stood%20up.%20He%20had%20shrunk%20considerably%20since%20I%E2%80%99d%20seen%20him%5Cnat%20the%20Bocuse%20d%E2%80%99Or.%20Then%20again%2C%20that%20last%20time%2C%20he%20had%20been%20in%20a%5Cntowering%20toque%20and%20heavy-heeled%20clogs%2C%20and%20doing%20that%20erect-chef-posture%5Cnroutine.%20Without%20the%20gear%2C%20he%20was%2C%20frankly%2C%20a%20little%20naked.%20He%20seemed%2C%5Cnto%20me%20at%20least%E2%80%94and%20it%20is%20startling%20for%20me%20even%20to%20utter%20the%20thought%E2%80%94to%5Cnbe%20almost%20normal.%20He%20was%20a%20man.%3C%2Fp%3E%5Cn%3Cp%3EThere%20was%20a%20lot%20about%20him%20that%20I%20didn%E2%80%99t%20know%2C%20I%20realized.%20Actually%2C%20on%5Cnreflection%2C%20I%20knew%20nothing.%20Or%2C%20at%20least%20now%E2%80%94alone%2C%20suddenly%2C%20with%20the%5Cndeity%20himself%E2%80%94I%20was%20feeling%20that%20I%20knew%20nothing.%20I%20hadn%E2%80%99t%20yet%20eaten%20at%5Cnhis%20main%20restaurant%2C%20L%E2%80%99Auberge%E2%80%94which%20has%20been%20awarded%20three%20Michelin%5Cnstars%20every%20year%20since%201965%2C%20making%20it%20the%20longest-running%20three-star%5Cnestablishment%20in%20the%20%3Cem%3EGuide%20Michelin%E2%80%99s%3C%2Fem%3E%20history%E2%80%94and%20was%20now%20angry%20with%5Cnmyself%20that%20I%20hadn%E2%80%99t.%20I%20felt%20foolish%2C%20and%20with%20the%20great%20man%20shaking%20my%5Cnhand%E2%80%94among%20the%20greatest%20meetings%20in%20my%20life%E2%80%94I%20went%20instantly%20and%5Cnuncharacteristically%20shy.%20He%20told%20me%20to%20follow.%20%E2%80%9CViens%2C%E2%80%9D%20he%20said%20gently.%3C%2Fp%3E%5Cn%3Cp%3E%E2%80%9C%3Cem%3EMerci%2C%20chef%3C%2Fem%3E%2C%E2%80%9D%20I%20said%20weakly.%3C%2Fp%3E%5Cn%3Cp%3EHe%20wanted%20to%20give%20me%20a%20tour.%3C%2Fp%3E%5Cn%3Cp%3E%E2%80%9C%3Cem%3EViens%2C%E2%80%9D%3C%2Fem%3E%20he%20repeated.%3C%2Fp%3E%5Cn%3Cp%3E%E2%80%9C%3Cem%3EMerci%2C%20chef%3C%2Fem%3E.%E2%80%9D%3C%2Fp%3E%5Cn%3Cp%3ESince%20Bocuse%20knew%20everyone%20at%20Les%20Halles%2C%20and%20since%20it%20was%20an%20early%20hour%5Cnand%20very%20few%20shoppers%20had%20shown%20up%20yet%2C%20our%20slow%20passage%20through%20its%5Cnaisles%20was%20surprisingly%20intimate.%3C%2Fp%3E%5Cn%3Cp%3E%E2%80%9CHere%20the%20charcuterie%20is%20very%20good%2C%E2%80%9D%20he%20said%2C%20not%20unlike%20an%20owner%20of%20a%5Cnlarge%20estate%20showing%20off%20roses%20that%20were%20having%20a%20good%20year.%20%E2%80%9CThis%20is%5Cnchez%20Sibilia.%E2%80%9D%3C%2Fp%3E%5Cn%3Cp%3EViannay%2C%20standing%20nearby%2C%20whispered%2C%20%E2%80%9CYou%20will%20buy%20your%3Cinline-embed%20type%3D%5C%22analytics%5C%22%20meta%3D%5C%22%257B%2522type%2522%253A%2522analytics%2522%252C%2522url%2522%253A%2522final%2522%252C%2522width%2522%253A%2522%2522%252C%2522height%2522%253A%2522%2522%252C%2522caption%2522%253A%2522%2522%257D%5C%22%20ref%3D%5C%22%5C%22%3E%3C%2Finline-embed%3E%20charcuterie%5Cnnowhere%20else.%20%3Cem%3ED%E2%80%99accord%3C%2Fem%3E%3F%E2%80%9D%3C%2Fp%3E%5Cn%3Cp%3E%E2%80%9C%3Cem%3EOui%2C%20chef%3C%2Fem%3E%2C%E2%80%9D%20I%20whispered%20back.%3C%2Fp%3E%5Cn%3Cp%3EBocuse%20and%20Sibilia%2C%20an%20imposing%2C%20no-nonsense%20woman%2C%20did%20kiss-kissy%5Cnbonjours%20and%20produced%20slices%20of%20rosette%2C%20the%20dried%20local%20sausage%20that%5Cnhistorians%20believe%20originated%20with%20mortadella%20and%20which%20the%20Lyonnais%5Cncrave%20whenever%20they%20leave%20town.%20They%20both%20stood%20around%20me%20while%20I%20tasted%5Cnit%2C%20studying%20my%20mouth%2C%20waiting%20for%20a%20verdict%2C%20as%20if%20such%20a%20tasting%20were%5Cna%20matter%20of%20great%20importance.%20It%20was%20a%20shtick%2C%20of%20course%2C%20their%20journo%5CnP.R.%20routine.%20I%20knew%20that.%20They%20knew%20that%20I%20knew%20that.%20But%2C%20meanwhile%2C%5Cnthe%20overwhelming%20fact%20was%20this%3A%20Bocuse%2C%20%3Cem%3Ethe%3C%2Fem%3E%20Paul%20Bocuse%2C%20was%5Cnhand-feeding%20me%20a%20%3Cem%3Esaucisson%3C%2Fem%3E%2C%20and%20I%20was%20remarkably%20O.K.%20with%20that.%3C%2Fp%3E%5Cn%3Cp%3E%E2%80%9C%3Cem%3EMerci%2C%20chef%3C%2Fem%3E.%E2%80%9D%3C%2Fp%3E%5Cn%3Cp%3EAcross%20the%20aisle%20was%20Les%20Volailles%20Clugnet%2C%20a%20bird%20vender.%20The%5Cnproprietor%20(%E2%80%9CBonjour%2C%20Pierre%E2%80%9D)%2C%20on%20seeing%20Bocuse%2C%20handed%20over%20one%20of%20his%20white%5Cnbirds%2C%20unsolicited.%20Bocuse%20seized%20it%20between%20his%20two%20very%20large%5Cnhands%20and%20slapped%20it%20around%2C%20firmly%20but%20with%20affection%2C%20as%20though%5Cnhandling%20the%20puppy%20of%20a%20favorite%20hunting%20dog.%3C%2Fp%3E%5Cn%3Cp%3E%E2%80%9CThe%20best%20chicken%20is%20the%20white%20breed%20from%20Bresse%2C%E2%80%9D%20Bocuse%20said.%20%E2%80%9C%3Cem%3ETout%5Cnle%20monde%20sait%20%C3%A7a.%3C%2Fem%3E%E2%80%9D%20Everyone%20knows%20this.%3C%2Fp%3E%5Cn%3Cp%3E%E2%80%9C%3Cem%3EMerci%2C%20chef%3C%2Fem%3E.%E2%80%9D%3C%2Fp%3E%5Cn%3Cp%3EHe%20handed%20the%20bird%20to%20me.%20It%20was%20heavy.%20Full.%20(Unlike%20the%20United%20States%2C%5CnFrance%20doesn%E2%80%99t%20gut%20birds%20until%20they%20are%20about%20to%20be%20cooked.)%20I%20held%20on%5Cnto%20the%20chicken%20and%20looked%20at%20it%20with%20polite%20curiosity.%20I%20didn%E2%80%99t%20feel%5Cnconfident%20enough%20to%20slap%20it%20around.%20Bocuse%20then%20did%20a%20riff%20on%20its%5Cnfeatures%2C%20the%20red%20waddle%20draped%20across%20my%20thumb%2C%20the%20white%20feathers%2C%20the%5Cndangling%20blue%20feet%2C%20and%20the%20many%20qualities%20that%20make%20the%20%3Cem%3Epoulet%20de%5CnBresse%3C%2Fem%3E%20a%20rare%20and%20scrumptious%20creature.%3C%2Fp%3E%5Cn%3Cp%3E%E2%80%9C%3Cem%3EMerci%2C%20chef%3C%2Fem%3E.%E2%80%9D%3C%2Fp%3E%5Cn%3Cp%3EViannay%20whispered%2C%20%E2%80%9CLes%20Volailles%20Clugnet.%20Do%20you%20understand%3F%20The%20only%5Cnplace.%E2%80%9D%3C%2Fp%3E%5Cn%3Cp%3E%E2%80%9C%3Cem%3EOui%2C%20chef%3C%2Fem%3E%2C%E2%80%9D%20I%20whispered%20back.%3C%2Fp%3E%5Cn%3Cp%3EThen%2C%20like%20that%2C%20bang%3A%20Bocuse%20said%20goodbye.%3C%2Fp%3E%5Cn%3Cp%3ELeaving%20already%3F%20It%20was%E2%80%94I%20can%E2%80%99t%20deny%20it%E2%80%94a%20moment%20of%20shocking%5Cndisappointment.%20So%20soon%3F%20After%20such%20a%20long%20wait%3F%20I%20hadn%E2%80%99t%20even%20asked%20a%5Cnquestion.%3C%2Fp%3E%5Cn%3Cp%3E%E2%80%9C%3Cem%3E%C3%80%20la%20prochaine%20fois%3C%2Fem%3E%2C%E2%80%9D%20he%20said.%3C%2Fp%3E%5Cn%3Cp%3E%E2%80%9CMerci%2C%20chef%2C%E2%80%9D%20I%20said%20sadly%2C%20but%20then%20brightened.%20%3Cem%3E%C3%80%20la%20prochaine%5Cnfois%3C%2Fem%3E%E2%80%94until%20next%20time%3F%20Was%20he%20telling%20me%20that%20he%E2%80%99d%20be%20seeing%20me%20again%3F%3C%2Fp%3E%5Cn%22%7D","ref":""},["p","Until ",["inline-embed",{"type":"analytics","meta":"%7B%22type%22%3A%22analytics%22%2C%22url%22%3A%221%22%2C%22width%22%3A%22%22%2C%22height%22%3A%22%22%2C%22caption%22%3A%22%22%7D","ref":""}],"1985 or thereabouts, Paul Bocuse was a very famous chef. He was a\nhousehold name, with an image so widely disseminated that just about\neveryone could tell you what he looked like—the big lips, the toque, and\nholding a live ",["em","poulet de Bresse"]," in his arms, say, and petting its\nhead—even if few people had any idea what his food might be like. Then,\naround 1985 or so, Bocuse turned into an icon. One moment: famous guy at\nstove. Then: Pope of restaurant people. He became the undisputed\nemissary of the kitchen mission. He became Frenchness."],["p","It is not entirely clear how this happened, because (and Bocuse would\nhave agreed) there were, and are, more talented and savvy chefs around.\nHe never hosted a food show. Although a master of the photo op (hamming\nit up in drag, stripping down to show off his tattoo of a Bresse\nchicken, straddling his Harley Davidson), he rarely appeared on\ntelevision. He published cookbooks. None was a game-changer. Apart from\ntwo forays abroad—one in Japan (which failed) and another at Walt Disney\nWorld, in Florida (which thrives, run by his son Jérôme)—he never really\nfranchised his name. And yet, in the mysterious, ineffable way of these\nthings, Bocuse, who died on Saturday, at the age of ninety-one, had\nsomething that no other chef has in the same abundance: an undeniably\ninfectious culinary charisma. Bocuse is what people want a grand French\nchef to be."],["p","In 2009, I moved to Lyon with my family, to train to be a French chef\n(not the “grand” kind, but more like a garden variety kitchen Frenchie).\nWe stayed five years. Bocuse was everywhere and nowhere. I enrolled in a\ncooking school—",["em","the"]," cooking school, l’Institut Paul Bocuse—where just\nabout every day there was a rumor that “Monsieur Paul” (as he is\nreferred to locally) was going to make an appearance. He never did,\nalthough he regularly interrupted the instructors, calling them on their\ncell phones during class. I was given hundreds of recipes and expected\nto memorize them; they were mainly ",["em","his"]," recipes, I would eventually\nlearn. Around the city there were three murals devoted to his image,\nincluding one that was directly across the street from our apartment on\nthe Saone river: he was what I looked at drinking my morning coffee.\nBridges were named after him. Les Halles, the indoor market, was\nofficially Les Halles Paul Bocuse. And so, too, was the world’s biggest\nand most hysterical cooking competition, the biennial Bocuse d’Or."],["p","The event is organized like a World Cup (forty-eight nations competing,\neach represented by a team of two, a chef and a deputy) and run like a\ndog show, with dishes paraded through a stadium in front of twenty-six\njudges. It culminates in an awards ceremony that manages to use the\niconography of the Olympics (with a podium for gold, silver, and\nbronze), the Oscars (the “Bocuse d’Or” is a gold statuette in the image\nof Paul Bocuse), and a New York City bar mitzvah (strobe lights, loud\noom-pah-pah music, and a ceiling drop of gold confetti). It is “tacky”\nmeets “technique,” but, despite the tackiness, the “technique” is real:\non display is the flashiest, most accomplished food on the planet."],["p","I attended my first Bocuse d’Or during our first year in France. I\nshowed up at 9 ",["em",{"class":"small"},"A.M."]," on the first day of what would be a two-day competition. The chefs had been there since five. Each two-member\nteam, plus a local ",["em","commis"],", a slave-helper provided by l’Institut Paul\nBocuse, was crushed bumpingly into a mini-kitchen cubicle the size of a\nchanging room at a bad beach resort. The cubicles, side by side, were\nspread out in a row, from left to right. They represented the “event.”\nIt was more rodeo arena than theatre. Facing the action were bleachers.\nThe mood in the mini-kitchen cubicles was of adrenaline and stress and\nsweat. Every cook was intensely aware of a clock ticking. They were\nfocussed and very quiet. The bleachers were very not. There were\nthousands of seats, and they were already filled. Many spectators\nwere in regional costume (the Japanese as samurais, the Mexicans in\nsombreros) or draped in national flags (the Swedes, the Americans).\nThere was a very loud mariachi band, a very loud drum-and-bugle corps,\na very loud pit percussion team, some guys banging cymbals, and many\nmorons with stadium air horns. There wasn’t anything obvious to be\nthundering about: no eviscerating of live goats, no fist pumping, no\nchef standing up and saying, “Bam!” But the cheering and the chanting\nnever stopped."],["p","I felt uninformed and naïve—how could I have not known that this kind of\ncooking (hunched-over, tortured people manipulating little things in\nlittle ways) was a nationalist sport?—when, lo and behold, Bocuse\nappeared."],["p","He had turned up backstage and taken it upon himself—the toque, the\nwhites, in a collar that looked like a French flag—to walk across the\nfloor. He padded by me, softly and stiffly, imposingly tall in his chef\nposture, doing small papal waves, and seemed not to notice that a queue\nof competitors was forming behind him. They were Japanese, Korean,\nIndian, Brazilian, Norwegian . . . actually, I have no idea where they\nwere from, except that there were lots of them, and they seemed not to\nbe speaking either French or English. Only a moment before, they were\nlocked down",["inline-embed",{"type":"analytics","meta":"%7B%22type%22%3A%22analytics%22%2C%22url%22%3A%222%22%2C%22width%22%3A%22%22%2C%22height%22%3A%22%22%2C%22caption%22%3A%22%22%7D","ref":""}]," in their panicky prep, when one of them noticed that Bocuse\nwas in the house and abandoned his station. He seemed to then realize, Hey,\nit’s O.K. to follow the great man around—or at least he registered that\nno one was making him stop—and called to the others in their\nmini-cubicle kitchens, and urged them to join in. The line quickly\nbecame much longer, conga-style."],["p","No one really knew what to do next—you can only get away with doing the\nconga behind Bocuse for so long—when someone touched the great man\nand dropped out of the queue, satisfied by the contact. The next person\nin the queue touched a sleeve. Then it was a shoulder, the back of\nBocuse’s hand. An Asian chef was next up, and he seized Bocuse’s apron\nby the hem, released it, and held the hand that had done the seizing by\nits wrist, staring at it, shaking it, and screamed with wide-eyed\ninsanity as though his skin were burning. A cook fell to his knees, and\nkissed the ground where the chef had tread (which, I don’t know, call me\nprudish, seemed a little excessive)."],["p","It got grabby. The line lost its linear definition, and, just as it was\nseeming outright dangerous, Bocuse was gone. Fearing a mob moment,\nhandlers had emerged and ushered him through a backstage door."],["p","For the rest of the year, it became my mission to meet the great man\nproperly. I was told that I could reach him through his people. A wife,\na daughter, a son. I wrote them. I phoned. It was the daughter who\nfinally got back to me, and there was disdain in her reply: How dare you\nthink that you, a lowly writer guy, would have anything to say to\nMonsieur Bocuse that is even remotely worth his time?"],["p","Then I got lucky. I happened to be describing my frustrations in the bar\nof La Mère Brazier, which is among the restaurants where Bocuse\ntrained. Mathieu Viannay is the chef, and he was standing nearby. What I\ndidn’t know is that Viannay was also a member of an “inner circle.” Its\nmembers—basically, kick-ass chefs who understand the soul and suffering\nof Lyonnais cooking—were awarded permanent access to the great man."],["p","Viannay interrupted my conversation. “You want to meet Monsieur Paul?”"],["p","“Yes,” I said, “very much.”"],["p","“Come by tomorrow morning at seven. Do not be late.”"],["p","Viannay was waiting for me in his car. We crossed the Rhône River and\ninto the Third Arrondissement and in the direction of Les Halles de Lyon\nPaul Bocuse. I then spotted the great chef’s vehicle, a massive black\nJeep Wrangler (American), with all the doo-dahs, spotlights on the roof,\na winch on the front bumper, and wide off-road tires (Michelin, of\ncourse). It was parked slapdash across a sidewalk, in front of the\nmarket-hall entrance. The sight confirmed two Lyonnais rumors: one, that\nBocuse always had his early-morning coffee at Les Halles; and, two, that\nthe police knew his vehicle and never ticketed it."],["p","For our part, we used a conventional parking lot, paid for it, entered\nthe hall, and stopped by a shellfish vendor named Leon, who gave us a\nplatter of oysters. (“Leon is the one you buy your oysters from,”\nViannay told me. “",["em","D’accord?"],"”)"],["p","We made our way to Le Boulanger, an informal café opposite the stand of\nLa Mère Richard, the city’s famous cheese lady, where Viannay paused to\npoint out the quality of what was on display. (“This is where you will\nbuy your cheese. ",["em","D’accord"],"?”)"],["p","I looked across the aisle. There he was: Paul Bocuse, sitting by\nhimself, finishing a coffee."],["p","He was slumped slightly in his chair, and, wearing a black Pringle\ncotton shirt, a worker’s cotton jacket, black trousers, and sneakers, he\nlooked like a regional train conductor at the end of his shift. On\nspotting us, he stood up. He had shrunk considerably since I’d seen him\nat the Bocuse d’Or. Then again, that last time, he had been in a\ntowering toque and heavy-heeled clogs, and doing that erect-chef-posture\nroutine. Without the gear, he was, frankly, a little naked. He seemed,\nto me at least—and it is startling for me even to utter the thought—to\nbe almost normal. He was a man."],["p","There was a lot about him that I didn’t know, I realized. Actually, on\nreflection, I knew nothing. Or, at least now—alone, suddenly, with the\ndeity himself—I was feeling that I knew nothing. I hadn’t yet eaten at\nhis main restaurant, L’Auberge—which has been awarded three Michelin\nstars every year since 1965, making it the longest-running three-star\nestablishment in the ",["em","Guide Michelin’s"]," history—and was now angry with\nmyself that I hadn’t. I felt foolish, and with the great man shaking my\nhand—among the greatest meetings in my life—I went instantly and\nuncharacteristically shy. He told me to follow. “Viens,” he said gently."],["p","“",["em","Merci, chef"],",” I said weakly."],["p","He wanted to give me a tour."],["p","“",["em","Viens,”"]," he repeated."],["p","“",["em","Merci, chef"],".”"],["p","Since Bocuse knew everyone at Les Halles, and since it was an early hour\nand very few shoppers had shown up yet, our slow passage through its\naisles was surprisingly intimate."],["p","“Here the charcuterie is very good,” he said, not unlike an owner of a\nlarge estate showing off roses that were having a good year. “This is\nchez Sibilia.”"],["p","Viannay, standing nearby, whispered, “You will buy your",["inline-embed",{"type":"analytics","meta":"%7B%22type%22%3A%22analytics%22%2C%22url%22%3A%22final%22%2C%22width%22%3A%22%22%2C%22height%22%3A%22%22%2C%22caption%22%3A%22%22%7D","ref":""}]," charcuterie\nnowhere else. ",["em","D’accord"],"?”"],["p","“",["em","Oui, chef"],",” I whispered back."],["p","Bocuse and Sibilia, an imposing, no-nonsense woman, did kiss-kissy\nbonjours and produced slices of rosette, the dried local sausage that\nhistorians believe originated with mortadella and which the Lyonnais\ncrave whenever they leave town. They both stood around me while I tasted\nit, studying my mouth, waiting for a verdict, as if such a tasting were\na matter of great importance. It was a shtick, of course, their journo\nP.R. routine. I knew that. They knew that I knew that. But, meanwhile,\nthe overwhelming fact was this: Bocuse, ",["em","the"]," Paul Bocuse, was\nhand-feeding me a ",["em","saucisson"],", and I was remarkably O.K. with that."],["p","“",["em","Merci, chef"],".”"],["p","Across the aisle was Les Volailles Clugnet, a bird vender. The\nproprietor (“Bonjour, Pierre”), on seeing Bocuse, handed over one of his white\nbirds, unsolicited. Bocuse seized it between his two very large\nhands and slapped it around, firmly but with affection, as though\nhandling the puppy of a favorite hunting dog."],["p","“The best chicken is the white breed from Bresse,” Bocuse said. “",["em","Tout\nle monde sait ça."],"” Everyone knows this."],["p","“",["em","Merci, chef"],".”"],["p","He handed the bird to me. It was heavy. Full. (Unlike the United States,\nFrance doesn’t gut birds until they are about to be cooked.) I held on\nto the chicken and looked at it with polite curiosity. I didn’t feel\nconfident enough to slap it around. Bocuse then did a riff on its\nfeatures, the red waddle draped across my thumb, the white feathers, the\ndangling blue feet, and the many qualities that make the ",["em","poulet de\nBresse"]," a rare and scrumptious creature."],["p","“",["em","Merci, chef"],".”"],["p","Viannay whispered, “Les Volailles Clugnet. Do you understand? The only\nplace.”"],["p","“",["em","Oui, chef"],",” I whispered back."],["p","Then, like that, bang: Bocuse said goodbye."],["p","Leaving already? It was—I can’t deny it—a moment of shocking\ndisappointment. So soon? After such a long wait? I hadn’t even asked a\nquestion."],["p","“",["em","À la prochaine fois"],",” he said."],["p","“Merci, chef,” I said sadly, but then brightened. ",["em","À la prochaine\nfois"],"—until next time? Was he telling me that he’d be seeing me again?"]]],"canonicalUrl":"https://www.newyorker.com/culture/annals-of-gastronomy/the-first-time-i-met-paul-bocuse","categories":{"sections":[{"id":"590411aa6870945b4d262380","modelName":"category","collection":"categories","contributors":{},"hierarchy":[{"id":"590411aa6870945b4d262380","modelName":"category","collection":"categories","contributors":{},"hierarchy":[],"name":"Culture","parent":[],"photos":{},"root":[],"slug":"culture"},{"id":"5902ab3b07071543e751a992","modelName":"category","collection":"categories","contributors":{},"hierarchy":[],"name":"Sections","parent":[],"photos":{},"root":[],"slug":"sections"}],"name":"Culture","parent":[{"id":"5902ab3b07071543e751a992","modelName":"category","collection":"categories","contributors":{},"hierarchy":[],"name":"Sections","parent":[],"photos":{},"root":[],"slug":"sections"}],"photos":{},"root":{"id":"5902ab3b07071543e751a992","modelName":"category","collection":"categories","contributors":{},"hierarchy":[],"name":"Sections","parent":[],"photos":{},"root":[],"slug":"sections"},"slug":"culture","children":[{"id":"59de4687640f52066c8c1b64","modelName":"category","collection":"categories","contributors":{},"hierarchy":[{"id":"59de4687640f52066c8c1b64","modelName":"category","collection":"categories","contributors":{},"hierarchy":[],"name":"Annals of Gastronomy","parent":[],"photos":{},"root":[],"slug":"annals-of-gastronomy"},{"id":"590411aa6870945b4d262380","modelName":"category","collection":"categories","contributors":{},"hierarchy":[],"name":"Culture","parent":[],"photos":{},"root":[],"slug":"culture"},{"id":"5902ab3b07071543e751a992","modelName":"category","collection":"categories","contributors":{},"hierarchy":[],"name":"Sections","parent":[],"photos":{},"root":[],"slug":"sections"}],"name":"Annals of Gastronomy","parent":[{"id":"590411aa6870945b4d262380","modelName":"category","collection":"categories","contributors":{},"hierarchy":[{"id":"590411aa6870945b4d262380","modelName":"category","collection":"categories","contributors":{},"hierarchy":[],"name":"Culture","parent":[],"photos":{},"root":[],"slug":"culture"},{"id":"5902ab3b07071543e751a992","modelName":"category","collection":"categories","contributors":{},"hierarchy":[],"name":"Sections","parent":[],"photos":{},"root":[],"slug":"sections"}],"name":"Culture","parent":[{"id":"5902ab3b07071543e751a992","modelName":"category","collection":"categories","contributors":{},"hierarchy":[],"name":"Sections","parent":[],"photos":{},"root":[],"slug":"sections"}],"photos":{},"root":{"id":"5902ab3b07071543e751a992","modelName":"category","collection":"categories","contributors":{},"hierarchy":[],"name":"Sections","parent":[],"photos":{},"root":[],"slug":"sections"},"slug":"culture"}],"photos":{},"root":{"id":"5902ab3b07071543e751a992","modelName":"category","collection":"categories","contributors":{},"hierarchy":[],"name":"Sections","parent":[],"photos":{},"root":[],"slug":"sections"},"slug":"annals-of-gastronomy"},{"id":"59c2827ea07dab556b6c7342","modelName":"category","collection":"categories","contributors":{},"hierarchy":[{"id":"59c2827ea07dab556b6c7342","modelName":"category","collection":"categories","contributors":{},"hierarchy":[],"name":"Postscript","parent":[],"photos":{},"root":[],"slug":"postscript"},{"id":"590411aa6870945b4d262380","modelName":"category","collection":"categories","contributors":{},"hierarchy":[],"name":"Culture","parent":[],"photos":{},"root":[],"slug":"culture"},{"id":"5902ab3b07071543e751a992","modelName":"category","collection":"categories","contributors":{},"hierarchy":[],"name":"Sections","parent":[],"photos":{},"root":[],"slug":"sections"}],"name":"Postscript","parent":[{"id":"590411aa6870945b4d262380","modelName":"category","collection":"categories","contributors":{},"hierarchy":[{"id":"590411aa6870945b4d262380","modelName":"category","collection":"categories","contributors":{},"hierarchy":[],"name":"Culture","parent":[],"photos":{},"root":[],"slug":"culture"},{"id":"5902ab3b07071543e751a992","modelName":"category","collection":"categories","contributors":{},"hierarchy":[],"name":"Sections","parent":[],"photos":{},"root":[],"slug":"sections"}],"name":"Culture","parent":[{"id":"5902ab3b07071543e751a992","modelName":"category","collection":"categories","contributors":{},"hierarchy":[],"name":"Sections","parent":[],"photos":{},"root":[],"slug":"sections"}],"photos":{},"root":{"id":"5902ab3b07071543e751a992","modelName":"category","collection":"categories","contributors":{},"hierarchy":[],"name":"Sections","parent":[],"photos":{},"root":[],"slug":"sections"},"slug":"culture"}],"photos":{},"root":{"id":"5902ab3b07071543e751a992","modelName":"category","collection":"categories","contributors":{},"hierarchy":[],"name":"Sections","parent":[],"photos":{},"root":[],"slug":"sections"},"slug":"postscript"}]}]},"channel":"","contentSource":"web","contentTags":[{"name":"Food","url":"/tag/food"},{"name":"Postscript","url":"/tag/postscript"},{"name":"Cooking","url":"/tag/cooking"},{"name":"France","url":"/tag/france"}],"contentType":"article","contributors":[{"name":"Bill Buford","type":"author","title":"Bill Buford’s next book is “Dirt (Five Years in the Ancient Kitchens of Lyon, Where My Children Became Frenchies and I Trained to be a Chef),” which will be published in September. A former fiction editor at \u003cem>The New Yorker\u003c/em>, he is also the author of “Among the Thugs” and “Heat.”","email":"","bio":null,"social":[],"url":"/contributors/bill-buford","image":null}],"dek":"","displayMostPopular":true,"formatting":{},"hed":"The First Time I Met Paul Bocuse","id":"5a6602e016b0267062e903f3","inlineEmbeds":{},"interactiveOverride":"","modifiedAt":"2018-01-23T00:37:42.779Z","photos":{"social":[{"id":"5a6633cd1ccea91635374116","modelName":"photo","collection":"photos","aspectRatios":{"2:2":{"width":3549,"height":3549,"override":false,"modifications":{"crop":{"x":1864,"y":1362,"height":3549,"width":3549}},"duration":null},"16:9":{"width":5470,"height":3076,"override":false,"modifications":{"crop":{"x":944,"y":1833,"height":3076,"width":5470}},"duration":null},"4:3":{"width":4716,"height":3537,"override":false,"modifications":{"crop":{"x":1381,"y":1374,"height":3537,"width":4716}},"duration":null},"1:1":{"width":3685,"height":3685,"override":false,"modifications":{"crop":{"x":1804,"y":1226,"height":3685,"width":3685}},"duration":null},"master":{"format":"JPEG","url":"https://cn-copilot-media.s3.amazonaws.com/public/tny-services/production/2018/01/22/5a6633cc1ccea91635374115_buford-paul-bocuse-2.jpg","width":7360,"height":4912,"duration":null}},"caption":["div",["p","Paul Bocuse, who died on Saturday, at the age of ninety-one, had an undeniably infectious culinary charisma. He was what people want a grand French chef to be."]],"altText":"","credit":"Photograph by Jeff Pachoud / AFP / Getty","filename":"buford-paul-bocuse-2.jpg","revision":6,"tags":[],"title":"","variants":{}}],"tout":[{"id":"5a6633cd1ccea91635374116","modelName":"photo","collection":"photos","aspectRatios":{"2:2":{"width":3549,"height":3549,"override":false,"modifications":{"crop":{"x":1864,"y":1362,"height":3549,"width":3549}},"duration":null},"16:9":{"width":5470,"height":3076,"override":false,"modifications":{"crop":{"x":944,"y":1833,"height":3076,"width":5470}},"duration":null},"4:3":{"width":4716,"height":3537,"override":false,"modifications":{"crop":{"x":1381,"y":1374,"height":3537,"width":4716}},"duration":null},"1:1":{"width":3685,"height":3685,"override":false,"modifications":{"crop":{"x":1804,"y":1226,"height":3685,"width":3685}},"duration":null},"master":{"format":"JPEG","url":"https://cn-copilot-media.s3.amazonaws.com/public/tny-services/production/2018/01/22/5a6633cc1ccea91635374115_buford-paul-bocuse-2.jpg","width":7360,"height":4912,"duration":null}},"caption":["div",["p","Paul Bocuse, who died on Saturday, at the age of ninety-one, had an undeniably infectious culinary charisma. He was what people want a grand French chef to be."]],"altText":"","credit":"Photograph by Jeff Pachoud / AFP / Getty","filename":"buford-paul-bocuse-2.jpg","revision":6,"tags":[],"title":"","variants":{}}],"lede":[{"id":"5a6633cd1ccea91635374116","modelName":"photo","collection":"photos","aspectRatios":{"2:2":{"width":3549,"height":3549,"override":false,"modifications":{"crop":{"x":1864,"y":1362,"height":3549,"width":3549}},"duration":null},"16:9":{"width":5470,"height":3076,"override":false,"modifications":{"crop":{"x":944,"y":1833,"height":3076,"width":5470}},"duration":null},"4:3":{"width":4716,"height":3537,"override":false,"modifications":{"crop":{"x":1381,"y":1374,"height":3537,"width":4716}},"duration":null},"1:1":{"width":3685,"height":3685,"override":false,"modifications":{"crop":{"x":1804,"y":1226,"height":3685,"width":3685}},"duration":null},"master":{"format":"JPEG","url":"https://cn-copilot-media.s3.amazonaws.com/public/tny-services/production/2018/01/22/5a6633cc1ccea91635374115_buford-paul-bocuse-2.jpg","width":7360,"height":4912,"duration":null}},"caption":["div",["p","Paul Bocuse, who died on Saturday, at the age of ninety-one, had an undeniably infectious culinary charisma. He was what people want a grand French chef to be."]],"altText":"","credit":"Photograph by Jeff Pachoud / AFP / Getty","filename":"buford-paul-bocuse-2.jpg","revision":6,"tags":[],"title":"","variants":{}}]},"promoDek":"","promoHed":"The First Time I Met Paul Bocuse","pubDate":"7:08 P.M.","related":{"relatedList":[{"author":{"name":"Priya Krishna","url":"/contributors/priya-krishna"},"dek":"Urvashi Pitre’s most hard-earned fans are her fellow-Indians, who tend to  be skeptical that a dish whipped up in fifteen minutes could qualify as authentic.","hed":"The “Butter-Chicken Lady” Who Made Indian Cooks Love the Instant Pot","id":"5a565d23a363b86dd5ec1b27","image":{"id":"5a57983910a8dc03151e8749","modelName":"photo","collection":"photos","aspectRatios":{"2:2":{"width":1199,"height":1199,"override":false,"modifications":{"crop":{"x":0,"y":183,"height":1199,"width":1199}},"duration":null},"16:9":{"width":1199,"height":674,"override":false,"modifications":{"crop":{"x":0,"y":689,"height":674,"width":1199}},"duration":null},"4:3":{"width":1181,"height":886,"override":false,"modifications":{"crop":{"x":18,"y":551,"height":886,"width":1181}},"duration":null},"1:1":{"width":1199,"height":1199,"override":false,"modifications":{"crop":{"x":0,"y":186,"height":1199,"width":1199}},"duration":null},"master":{"format":"JPEG","url":"https://cn-copilot-media.s3.amazonaws.com/public/tny-services/production/2018/01/11/5a57983910a8dc03151e8748_Krishna-Butter-Chicken-Lady-Instant-Pot.jpg","width":1200,"height":1474,"duration":null}},"caption":"Urvashi Pitre’s Instant Post recipe for butter chicken earned her a following for her easy versions of classic Indian recipes.","altText":"","credit":"Photograph by Melina Hammer / Courtesy Callisto Media","filename":"Krishna-Butter-Chicken-Lady-Instant-Pot.jpg","revision":4,"tags":[],"title":"","variants":{}},"publishDate":"Jan. 11, 2018","rubric":{"name":"Annals of Gastronomy","url":"/culture/annals-of-gastronomy"},"source":"curated","url":"/culture/annals-of-gastronomy/the-butter-chicken-lady-who-made-indian-cooks-love-the-instant-pot"},{"author":{"name":"Gustavo Arellano","url":"/contributors/gustavo-arellano"},"dek":"Recent Mexican immigrants deride them as a gringo quirk. Foodie purists dismiss them as not “real” Mexican food. But they can be revelatory.","hed":"In Praise of Flour Tortillas, an Unsung Jewel of the U.S.-Mexico Borderlands","id":"5a5797f612d6f43d2a07d023","image":{"id":"5a57e3f9eee8522c37ed4b2b","modelName":"photo","collection":"photos","aspectRatios":{"2:2":{"width":730,"height":730,"override":false,"modifications":{"crop":{"x":410,"y":0,"height":730,"width":730}},"duration":null},"16:9":{"width":1200,"height":675,"override":false,"modifications":{"crop":{"x":0,"y":61,"height":675,"width":1200}},"duration":null},"4:3":{"width":1063,"height":797,"override":false,"modifications":{"crop":{"x":69,"y":2,"height":797,"width":1063}},"duration":null},"1:1":{"width":717,"height":717,"override":false,"modifications":{"crop":{"x":404,"y":0,"height":717,"width":717}},"duration":null},"master":{"format":"JPEG","url":"https://cn-copilot-media.s3.amazonaws.com/public/tny-services/production/2018/01/11/5a57e3f9eee8522c37ed4b2a_Arellano-In-Praise-of-Flour-Tortillas.jpg","width":1200,"height":800,"duration":null}},"caption":"Recent Mexican immigrants deride them as a gringo quirk. Foodie purists dismiss them as not “real” Mexican food. But good flour tortillas can be revelatory.","altText":"","credit":"Photograph by YinYang / Getty \n","filename":"Arellano-In-Praise-of-Flour-Tortillas.jpg","revision":6,"tags":[],"title":"","variants":{}},"publishDate":"Jan. 13, 2018","rubric":{"name":"Annals of Gastronomy","url":"/culture/annals-of-gastronomy"},"source":"curated","url":"/culture/annals-of-gastronomy/in-praise-of-flour-tortillas-an-unsung-jewel-of-the-us-mexico-borderlands"}],"featuredVideo":[{"description":"Domenico DeMarco has been making pizzas at Brooklyn’s famous Di Fara Pizza since 1965.","photo":{"url":"http://dwgyu36up6iuz.cloudfront.net/heru80fdn/image/upload/c_fill,d_placeholder_thescene.jpg,fl_progressive,g_center,h_360,q_80,w_640/v1510335702/thenewyorker_a-priceless-pizzeria-in-brooklyn.jpg"},"title":"A Priceless Pizzeria in Brooklyn","videoData":{"id":"5a05e23a24b6db64c8470132","scriptEmbedUrl":"//player.cnevids.com/script/video/5a0345662d1ca0524d000003.js?autoplay=0"}}]},"rubric":{"name":"Annals of Gastronomy","url":"/culture/annals-of-gastronomy"},"seoTitle":"The First Time I Met Paul Bocuse","seoDescription":"Bill Buford remembers the French chef Paul Bocuse, who died on Saturday, at the age of ninety-one.","socialDescription":"Bocuse, who died on Saturday, at the age of ninety-one, had an undeniably infectious culinary charisma. He was what people want a grand French\nchef to be. ","socialTitle":"The First Time I Met Paul Bocuse","subChannel":"","tags":["food","postscript","cooking","france"],"template":"default","url":"/culture/annals-of-gastronomy/the-first-time-i-met-paul-bocuse","videos":{},"wordCount":2150,"isoPubDate":"2018-01-23T00:08:28.877Z","social":{"facebook":{"name":"facebook","shareLink":"https://www.facebook.com/sharer/sharer.php?u=https%3A%2F%2Fwww.newyorker.com%2Fculture%2Fannals-of-gastronomy%2Fthe-first-time-i-met-paul-bocuse&display=popup&ref=plugin","svg":"\u003csvg width=\"8\" height=\"17\" viewBox=\"0 0 8 17\" xmlns=\"http://www.w3.org/2000/svg\">\u003cpath d=\"M1.7323 17H5.2V8.499h2.3135L7.82 5.5698H5.2l.0037-1.4667c0-.7634.072-1.1733 1.154-1.1733h1.4458V0h-2.314C2.7104 0 1.7324 1.4214 1.7324 3.811v1.7588H0v2.9297h1.7323V17z\" fill=\"#000\" fill-rule=\"evenodd\"/>\u003c/svg>"},"twitter":{"name":"twitter","shareLink":"https://twitter.com/intent/tweet?original_referer=https%3A%2F%2Fwww.newyorker.com%2Fculture%2Fannals-of-gastronomy%2Fthe-first-time-i-met-paul-bocuse&text=The%20First%20Time%20I%20Met%20Paul%20Bocuse&tw_p=tweetbutton&url=https%3A%2F%2Fwww.newyorker.com%2Fculture%2Fannals-of-gastronomy%2Fthe-first-time-i-met-paul-bocuse","svg":"\u003csvg width=\"19\" height=\"15\" viewBox=\"0 0 19 15\" xmlns=\"http://www.w3.org/2000/svg\">\u003cpath d=\"M18.223 1.739c-.654.295-1.3565.4943-2.094.584.7527-.459 1.3308-1.1858 1.603-2.0518-.7045.425-1.4847.7336-2.3152.9C14.7518.4504 13.8043 0 12.7558 0c-2.0137 0-3.646 1.6605-3.646 3.7084 0 .2908.032.5737.0943.8452-3.03-.1547-5.7164-1.631-7.5146-3.8748-.3138.5478-.4936 1.1848-.4936 1.8644 0 1.2866.6436 2.4217 1.622 3.0868-.5977-.0192-1.16-.186-1.6515-.464l-.0003.0468c0 1.7968 1.2567 3.2957 2.9246 3.6364-.306.0847-.628.13-.9605.13-.235 0-.4632-.0232-.686-.0665.464 1.4733 1.8105 2.5455 3.406 2.5755-1.248.9946-2.82 1.5875-4.528 1.5875-.2943 0-.5845-.0176-.8698-.0518 1.6136 1.052 3.53 1.666 5.5888 1.666 6.706 0 10.3733-5.651 10.3733-10.5515 0-.1608-.0036-.3208-.0106-.4798.7123-.5228 1.3304-1.176 1.8192-1.9197\" fill=\"#000\" fill-rule=\"evenodd\"/>\u003c/svg>"},"email":{"name":"email","shareLink":"mailto:?subject=From%20newyorker.com:%20The%20First%20Time%20I%20Met%20Paul%20Bocuse&body=Bocuse%2C%20who%20died%20on%20Saturday%2C%20at%20the%20age%20of%20ninety-one%2C%20had%20an%20undeniably%20infectious%20culinary%20charisma.%20He%20was%20what%20people%20want%20a%20grand%20French%0Achef%20to%20be.%20%0D%0Ahttps%3A%2F%2Fwww.newyorker.com%2Fculture%2Fannals-of-gastronomy%2Fthe-first-time-i-met-paul-bocuse","svg":"\u003csvg width=\"19\" height=\"14\" viewBox=\"0 0 19 14\" xmlns=\"http://www.w3.org/2000/svg\">\u003cpath d=\"M9.7902 9.4853l-.003-.003-.004.003-6.9808-7.071 1.396-1.414 5.5888 5.66 5.588-5.66 1.396 1.414-6.981 7.071zm-8.9624 4.515h17.7707v-14H.8278v14z\" fill=\"#000\" fill-rule=\"evenodd\"/>\u003c/svg>"},"print":{"name":"print","shareLink":"#","svg":"\u003csvg width=\"17\" height=\"17\" viewBox=\"0 0 17 17\" xmlns=\"http://www.w3.org/2000/svg\">\u003cg fill=\"#FFF\" fill-rule=\"evenodd\">\u003cpath d=\"M4.9688 1.988h8.008v2.009h.999V1.988c0-.559-.456-.999-.999-.999h-8.008c-.559 0-.999.456-.999.999v2.009h.999V1.988zM14.9857 8.0037c-.559 0-.999-.456-.999-.999 0-.559.456-.999.999-.999.559 0 .999.456.999.999.016.543-.44.999-.999.999zm-1.993 8.008h-8.024v-7.009h8.008v7.015h.016v-.006zm3.008-11.016H1.9777c-.56 0-.999.456-.999.999v5.001c0 .559.456.998.999.998h2.008v4.007c0 .559.456.999.999.999h8.008c.56 0 .999-.456.999-.999v-3.985h2.009c.559 0 .999-.456.999-.999v-5.022c0-.543-.456-.999-.999-.999z\"/>\u003cpath d=\"M5.979 11.012h4.007v-.999H5.979M5.979 13.01h6.016v-.999H5.979M5.979 15.018h6.016v-.999H5.979\"/>\u003c/g>\u003c/svg>"},"copyLink":{"name":"copyLink","shareLink":"https://www.newyorker.com/culture/annals-of-gastronomy/the-first-time-i-met-paul-bocuse","svg":"\u003csvg width=\"20px\" height=\"20px\" viewBox=\"0 0 20 20\" version=\"1.1\" xmlns=\"http://www.w3.org/2000/svg\" xmlns:xlink=\"http://www.w3.org/1999/xlink\"> \u003cg id=\"design\" stroke=\"none\" stroke-width=\"1\" fill=\"none\" fill-rule=\"evenodd\"> \u003cg id=\"current-1280\" transform=\"translate(-506.000000, -581.000000)\" fill=\"#000000\"> \u003cg id=\"BODY\" transform=\"translate(0.000000, 512.000000)\"> \u003cg id=\"SYRIA\" transform=\"translate(250.000000, 0.000000)\"> \u003cg id=\"hed\"> \u003cg id=\"LINK-SHARE\" transform=\"translate(11.000000, 66.000000)\"> \u003cg id=\"link\" transform=\"translate(243.000000, 1.000000)\"> \u003cg id=\"Group-7\" transform=\"translate(2.000000, 2.000000)\"> \u003cpath d=\"M12.143,7.8787 C11.901,7.6367 11.64,7.4247 11.364,7.2427 L9.891,8.7157 C10.193,8.8587 10.48,9.0437 10.729,9.2927 C11.899,10.4627 11.899,12.3657 10.729,13.5357 L7.9,16.3637 C6.731,17.5337 4.827,17.5337 3.658,16.3637 C2.488,15.1947 2.488,13.2907 3.658,12.1217 L5.072,10.7067 L3.658,9.2927 L2.244,10.7067 C0.294,12.6567 0.294,15.8287 2.244,17.7777 C4.193,19.7277 7.365,19.7277 9.315,17.7777 L12.143,14.9497 C14.092,13.0007 14.092,9.8277 12.143,7.8787\" id=\"Fill-3\">\u003c/path> \u003cpath d=\"M7.9004,12.1213 C8.1424,12.3633 8.4034,12.5753 8.6794,12.7573 L10.1524,11.2843 C9.8504,11.1413 9.5634,10.9563 9.3144,10.7073 C8.1444,9.5373 8.1444,7.6343 9.3144,6.4643 L12.1434,3.6363 C13.3124,2.4663 15.2164,2.4663 16.3854,3.6363 C17.5554,4.8053 17.5554,6.7093 16.3854,7.8783 L14.9714,9.2933 L16.3854,10.7073 L17.8004,9.2933 C19.7494,7.3433 19.7494,4.1713 17.8004,2.2213 C15.8504,0.2723 12.6784,0.2723 10.7284,2.2213 L7.9004,5.0503 C5.9514,6.9993 5.9514,10.1723 7.9004,12.1213\" id=\"Fill-5\">\u003c/path> \u003c/g> \u003c/g> \u003c/g> \u003c/g> \u003c/g> \u003c/g> \u003c/g> \u003c/g>\u003c/svg>"}},"externalSyndicationCanonicalUrl":"","syndicationCanonicalUrl":"https://www.newyorker.com/culture/annals-of-gastronomy/the-first-time-i-met-paul-bocuse","publishTimestamp":"7:08 P.M.","modifiedAtTimestamp":"2018-01-23T00:37:42.779Z"},"secondary":{"related":[],"trending":[{"author":{"name":"Adam Entous","type":"author","url":"/contributors/adam-entous"},"contributors":[{"name":"Adam Entous","type":"author","url":"/contributors/adam-entous"},{"name":"Evan Osnos","type":"author","url":"/contributors/evan-osnos"}],"dek":"China’s suspect courtship with Jared Kushner.","hed":"Jared Kushner Is China’s Trump Card","id":"5a628198470b7f16df06a9f9","image":{"cdn":"https://media.newyorker.com","collection":"photos","id":"5a625c6f64be70121c1040ea","filename":"180129_r31383.jpg","modelName":"photo"},"issueDate":{"name":"January 29, 2018","url":"/magazine/2018/01/29"},"publishDate":"","rubric":{"name":"The Political Scene"},"shouldLazyLoad":false,"source":"parsely","url":"/magazine/2018/01/29/jared-kushner-is-chinas-trump-card"},{"author":{"name":"Kathryn Schulz","type":"author","url":"/contributors/kathryn-schulz"},"contributors":[{"name":"Kathryn Schulz","type":"author","url":"/contributors/kathryn-schulz"}],"dek":"A brilliant début launched the career of a singular talent. How did he get lost?","hed":"William Melvin Kelley: The Lost Giant of American Literature","id":"5a5f731a1628db2787fece38","image":{"cdn":"https://media.newyorker.com","collection":"photos","id":"5a60ee09470b7f16df06a8d3","filename":"180129_r31345.jpg","modelName":"photo"},"issueDate":{"name":"January 29, 2018","url":"/magazine/2018/01/29"},"publishDate":"","rubric":{"name":"American Archives"},"shouldLazyLoad":false,"source":"parsely","url":"/magazine/2018/01/29/the-lost-giant-of-american-literature"},{"author":{"name":"Benjamin Wallace-Wells","type":"author","url":"/contributors/benjamin-wallace-wells"},"contributors":[{"name":"Benjamin Wallace-Wells","type":"author","url":"/contributors/benjamin-wallace-wells"}],"dek":"The Democrats have discovered the limits of their power in Trump’s Washington.","hed":"Did Schumer Cave on the Shutdown?","id":"5a6640de72667810f99bcf69","image":{"cdn":"https://media.newyorker.com","collection":"photos","id":"5a66264d44377923a90ad220","filename":"BWW-Deal-Reopen-Government.jpg","modelName":"photo"},"issueDate":{},"publishDate":"Jan. 22, 2018","rubric":{"name":"News Desk"},"shouldLazyLoad":true,"source":"parsely","url":"/news/news-desk/did-schumer-cave"},{"author":{"name":"Nick Paumgarten","type":"author","url":"/contributors/nick-paumgarten"},"contributors":[{"name":"Nick Paumgarten","type":"author","url":"/contributors/nick-paumgarten"}],"dek":"A movie made in a prison captures inmates’ hope and despair.","hed":"A Prison Film Made in Prison","id":"5a5f732e6829117ce32219be","image":{"cdn":"https://media.newyorker.com","collection":"photos","id":"5a610af802238d157101ce57","filename":"180129_r31358.jpg","modelName":"photo"},"issueDate":{"name":"January 29, 2018","url":"/magazine/2018/01/29"},"publishDate":"","rubric":{"name":"A Reporter at Large"},"shouldLazyLoad":true,"source":"parsely","url":"/magazine/2018/01/29/a-prison-film-made-in-prison"},{"author":{"name":"Ian Crouch","type":"author","url":"/contributors/ian-crouch"},"contributors":[{"name":"Ian Crouch","type":"author","url":"/contributors/ian-crouch"}],"dek":"There’s plenty to dislike about the New England Patriots, as people remind each other annually at this point in the season.","hed":"The Patriots’ Marvellous Comeback Against the Jaguars, and the Muddled Joy of Rooting for the Bad Guys","id":"5a6609e13085ef230b29d90f","image":{"cdn":"https://media.newyorker.com","collection":"photos","id":"5a66162744377923a90ad198","filename":"Crouch-Patriots-Marvelous-Comeback-Against-Jaguars.jpg","modelName":"photo"},"issueDate":{},"publishDate":"Jan. 22, 2018","rubric":{"name":"Sporting Scene"},"shouldLazyLoad":true,"source":"parsely","url":"/news/sporting-scene/the-patriots-marvelous-comeback-against-the-jaguars-and-the-muddled-joy-of-rooting-for-the-bad-guys"},{"author":{"name":"T. Coraghessan Boyle","type":"author","url":"/contributors/t-coraghessan-boyle"},"contributors":[{"name":"T. Coraghessan Boyle","type":"author","url":"/contributors/t-coraghessan-boyle"}],"dek":"Houses vanished, trees were uprooted like weeds, and the slurry of mud and ash rose as high as fifteen feet in some places.","hed":"After the Mudslides, an Absence in Montecito","id":"5a661e2a0b6b8326a5ec9982","image":{"cdn":"https://media.newyorker.com","collection":"photos","id":"5a62629d16b0267062e90097","filename":"Boyle-California-Mudslides.jpg","modelName":"photo"},"issueDate":{},"publishDate":"Jan. 22, 2018","rubric":{"name":"Cultural Comment"},"shouldLazyLoad":true,"source":"parsely","url":"/culture/cultural-comment/after-the-mudslides-an-absence-in-montecito"},{"author":{"name":"Amanda Fortini","type":"author","url":"/contributors/amanda-fortini"},"contributors":[{"name":"Amanda Fortini","type":"author","url":"/contributors/amanda-fortini"}],"dek":"The rally exposed lingering political divides in the anti-Trump movement.","hed":"Fault Lines at the National Women’s March, in Las Vegas","id":"5a66757c46f0914f553c611c","image":{"cdn":"https://media.newyorker.com","collection":"photos","id":"5a66899c4317dc20fb22e209","filename":"Fortini-Faultlines-National-Womens-March-Las-Vegas.jpg","modelName":"photo"},"issueDate":{},"publishDate":"Jan. 23, 2018","rubric":{"name":"News Desk"},"shouldLazyLoad":true,"source":"parsely","url":"/news/news-desk/fault-lines-at-the-national-womens-march-in-las-vegas"},{"author":{"name":"Kristen Roupenian","type":"author","url":"/contributors/kristen-roupenian"},"contributors":[{"name":"Kristen Roupenian","type":"author","url":"/contributors/kristen-roupenian"}],"dek":"“It was a terrible kiss, shockingly bad; Margot had trouble believing that a grown man could possibly be so bad at kissing.”","hed":"“Cat Person”","id":"5a1ed80b4f6fb117ea84ffc7","image":{"cdn":"https://media.newyorker.com","collection":"photos","id":"5a1ee51a45f728682b6d351c","filename":"171211_r31104web.jpg","modelName":"photo"},"issueDate":{"name":"December 11, 2017","url":"/magazine/2017/12/11"},"publishDate":"","rubric":{"name":"Fiction"},"shouldLazyLoad":true,"source":"parsely","url":"/magazine/2017/12/11/cat-person"},{"author":{"name":"Amy Davidson Sorkin","type":"author","url":"/contributors/amy-davidson-sorkin"},"contributors":[{"name":"Amy Davidson Sorkin","type":"author","url":"/contributors/amy-davidson-sorkin"}],"dek":"The President is headed to Davos. How can someone so “America first” attend an event so globally minded?","hed":"President Trump Goes to Davos","id":"5a62286897435c1de3ffea5f","image":{"cdn":"https://media.newyorker.com","collection":"photos","id":"5a62781164be70121c1040f2","filename":"180129_r31367illuweb-tout.jpg","modelName":"photo"},"issueDate":{"name":"January 29, 2018","url":"/magazine/2018/01/29"},"publishDate":"","rubric":{"name":"Comment"},"shouldLazyLoad":true,"source":"parsely","url":"/magazine/2018/01/29/trump-in-the-world-that-he-insults"},{"author":{"name":"John Cassidy","type":"author","url":"/contributors/john-cassidy"},"contributors":[{"name":"John Cassidy","type":"author","url":"/contributors/john-cassidy"}],"dek":"These protests represent a phenomenon that is more consequential than a shutdown: a rare popular mobilization against a sitting President.","hed":"The Women’s Marches Could Have More Lasting Consequences Than the Government Shutdown","id":"5a65e2e1b721560ea71e4933","image":{"cdn":"https://media.newyorker.com","collection":"photos","id":"5a660bfba7c43a0fe1e88b23","filename":"Cassidy-The-Womens-Marches.jpg","modelName":"photo"},"issueDate":{},"publishDate":"Jan. 22, 2018","rubric":{"name":"Our Columnists"},"shouldLazyLoad":true,"source":"parsely","url":"/news/our-columnists/the-womens-marches-could-have-more-lasting-consequences-than-the-government-shutdown"}]}};</script><script async="" src="/js/article.js?cb=8d41d241"></script><div><div id="parsely-root" style="display:none;"><div id="parsely-cfg" data-parsely-site="newyorker.com"></div></div><script>
  (function(s, p, d) {
    var h=d.location.protocol, i=p+"-"+s,
        e=d.getElementById(i), r=d.getElementById(p+"-root"),
        u=h==="https:"?"d1z2jf7jlzjs58.cloudfront.net"
        :"static."+p+".com";
    if (e) return;
    e = d.createElement(s); e.id = i; e.async = true;
    e.src = h+"//"+u+"/p.js"; r.appendChild(e);
  })("script", "parsely", document);
</script></div><script>window.digitalData = {"appVersion":"1.0","brand":"The New Yorker","canonical":"https://www.newyorker.com/culture/annals-of-gastronomy/the-first-time-i-met-paul-bocuse","contentID":"5a6602e016b0267062e903f3","contentLength":3,"contentSource":"web","contributor":"Bill Buford","display":"The First Time I Met Paul Bocuse","env":"production","keywords":{"display":"Food|Postscript|Cooking|France"},"magazineTOCSection":"","modifiedDate":"2018-01-23T00:37:42.779Z","numOfSearchResults":0,"pageType":"Post","pageValue":"page_1","publishDate":"2018-01-22T19:08:28-05:00","searchTerms":"","section":"Culture","subsection":"Annals of Gastronomy","syndication":"","syndicationUrl":""}</script><script defer="" src="https://assets.adobedtm.com/6372cf21ef88ee60bc2977a4898dcb5c7945a212/satelliteLib-fd2e38c2a1cccb723029dc3b607e196906c8bf93.js"></script><script async="" defer="" src="https://cdn.optimizely.com/js/2723621000.js"></script></body></html>
"""
```
